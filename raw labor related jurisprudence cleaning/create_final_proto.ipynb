# merge answers__df with juris__df.csv where the the file_path column and file_path column in both tables is equal 

```python
result = answers__df.merge(juris__df, how='inner', left_on='abs_file_path', right_on='file_path')
```

```python
result
```

# merge resulting df with juris.csv where the file_url column and identifier column is equal

```python
final = result.merge(juris, how='inner', left_on='identifier', right_on='file_url')
```

```python
final
```

```python
# delete identifier column and file_path column since they
# are essentially the same as the file_url column and the
# abs_file_path column
del final['identifier']
del final['file_path']


# rename also year_x, year_y, gr Number, and file_url columns
# to year_range, year, gr_number, and file_name respectively
final.rename(columns={'year_x': 'year_range', 'year_y': 'year', 'gr Number':'gr_number', 'file_url': 'file_name'}, inplace=True)
final
```

# reparse files and fill in missing values in these files
* division has 3375 missing values
    - examle of a dvision is first division, second division

```python
columns_with_null = []
for column in final.columns:

    null_rows = final[column].isnull()
    unique, counts = np.unique(null_rows, return_counts=True)

    # check if there are non values in column
    # if there is any count the number of nulls
    print(f'number of NaN in column {column}: {counts}')
    print(f'boolean values: {unique}')
    
    # extract all row indeces of each column with null values
    rows = final.loc[null_rows, :]
    columns_with_null.append(rows)
    print(rows)

null_df = dict(zip(final.columns, columns_with_null))
```
