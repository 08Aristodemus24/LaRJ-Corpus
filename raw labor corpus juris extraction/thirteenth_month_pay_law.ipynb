```python
page_text_content = [
    test.collect_content_individually_bs("""<div style="text-align: justify;">
    <strong>
    <font face="Verdana">
    <font color="#524744">
    <em>WHEREAS</em>,
    it is necessary to further protect the level of real wages from the
    ravage
    of worldwide inflation;</font>
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    <font color="#524744">
    <em>WHEREAS</em>,
    there has been no increase in the legal minimum wage rates since 1970;</font>
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    <font color="#524744">
    <em>WHEREAS</em>,
    the Christmas season is an opportune time for society to show its
    concern
    for the plight of the working masses so they may properly celebrate
    Christmas
    and New Year.</font>
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    <font color="#524744">
    <em>NOW,
    THEREFORE</em>,
    I, FERDINAND E. MARCOS, by virtue of the powers vested in me by the
    Constitution,
    do hereby decree as follows:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    <font color="#524744">Section 1.
    All employers are hereby required to pay all their employees receiving
    a basic salary of not more than P1,000 a month, regardless of the
    nature
    of their employment, a 13th-month pay not later than December 24 of
    every
    year.</font>
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    <font color="#524744">Sec. 2.
    Employers
    already paying their employees a 13th-month pay or its equivalent are
    not
    covered by this Decree.</font>
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    <font color="#524744">Sec. 3. This
    Decree shall take effect immediately.</font>
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    <font color="#524744">Done in the
    City of Manila, this 16th day of December 1975.</font>
    </font>
    </strong>
    <br>
    <strong>
    <font face="Verdana">
    </font>
    </strong>
    </div>"""),
    test.collect_content_individually_bs("""<center>
&nbsp;
<center>
<a name="PRESIDENTIAL DECREE NO. 851-RULES">
</a>
</center>
&nbsp;
<center>
<strong>
<tt>
<font color="#993300">
<font size="+1">PRESIDENTIAL
DECREE
NO. 851</font>
</font>
</tt>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#ff0000">&nbsp;</font>
</font>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#524744">RULES
AND
REGULATIONS IMPLEMENTING</font>
</font>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#524744">PRESIDENTIAL
DECREE NO. 851<br>
&nbsp;&nbsp; <br>
</font>
</font>
</strong>
</center>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">By virtue of the
powers vested in me by law, the following rules and regulations
implementing
<a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Presidential
Decree No. 851</a> are hereby issued for the guidance of all concerned.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Section 1.
<em>Payment of 13th-month Pay.</em> - All employers covered by <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Presidential
Decree No. 851</a>, hereinafter referred to as the <em>
<a href="https://www.chanrobles.com/presidentialdecreeno851.htm">"Decree"</a>
</em>,
shall pay to all their employees receiving a basic salary of not more
than
P1,000 a month a thirteenth-month pay not later than December 24 of
every
year.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Sec. 2. <em>Definition
of certain terms. </em>- As used in this issuance:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</strong>
</div>
<ul style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">(a) <em>"Thirteenth-month
pay"</em> shall mean one twelfth (1/12) of the basic salary of an
employee
within a calendar year;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(b) <em>"Basic
salary"</em> shall include all remunerations or earnings paid by an
employer
to an employee for services rendered but may not include cost-of-living
allowances granted pursuant to Presidential Decree No. 525 or Letter of
Instructions No. 174, profit-sharing payments, and all allowances and
monetary
benefits which are not considered or integrated as part of the regular
or basic salary of the employee at the time of the promulgation of the <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
on December 16, 1975.</font>
</font>
</strong>
</ul>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">Sec. 3. <em>Employers
covered.</em> - The <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
shall apply to all employers except to:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</strong>
</div>
<ul style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">(a)
Distressed
employers, such as (1) those which are currently incurring substantial
losses or (2) in the case of non-profit institutions and organizations,
where their income, whether from donations, contributions, grants and
other
earnings from any source, has consistently declined by more than forty
(40%) percent of their normal income for the last two (2) years,
subject
to the provision of Section 7 of this issuance;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(b) The
Government
and any of its political subdivisions, including government-owned and
controlled
corporations, except those corporations operating essentially as
private
subsidiaries of the Government;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(c) Employers
already paying their employees 13-month pay or more in a calendar year
or its equivalent at the time of this issuance;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(d) Employers
of household helpers and persons in the personal service of another in
relation to such workers; and</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(e) Employers
of those who are paid on purely commission, boundary, or task basis,
and
those who are paid a fixed amount for performing a specific work,
irrespective
of the time consumed in the performance thereof, except where the
workers
are paid on piece-rate basis in which case the employer shall be
covered
by this issuance insofar as such workers are concerned.</font>
</font>
</strong>
</ul>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">As used herein,
workers paid on piece-rate basis shall refer to those who are paid a
standard
amount for every piece or unit of work produced that is more or less
regularly
replicated, without regard to the time spent in producing the same.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">The term <em>"its
equivalent"</em> as used in paragraph c) hereof shall include Christmas
bonus, mid-year bonus, profit-sharing payments and other cash bonuses
amounting
to not less than 1/12th of the basic salary but shall not include cash
and stock dividends, cost of living allowances and all other allowances
regularly enjoyed by the employee, as well as non-monetary benefits.
Where
an employer pays less than 1/12th of the employees basic salary, the
employer
shall pay the difference.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Sec. 4. <em>Employees
covered.</em> - Except as provided in Section 3 of this issuance, all
employees
of covered employers shall be entitled to benefit provided under the <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
who are receiving not more than P1,000 a month, regardless of their
position,
designation or employment status, and irrespective of the method by
which
their wages are paid, provided that they have worked for at least one
month
during the calendar year.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Sec. 5. <em>Option
of covered employers.</em> - A covered employer may pay one-half of the
13th-month pay required by the <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
before the opening of the regular school year and the other half on or
before the 24th day of December of every year.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">In any
establishment
where a union has been recognized or certified as the collective
bargaining
agent of the employees therein, the periodicity or frequency of payment
of the 13th-month pay may be the subject of agreement.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Nothing herein
shall prevent employers from giving the benefits provided in the <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
to their employees who are receiving more than One Thousand (P1,000)
Pesos
a month or benefits higher than those provided by the <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Sec. 6. <em>Special
feature of benefit. </em>- The benefits granted under this issuance
shall
not be credited as part of the regular wage of the employees for
purposes
of determining overtime and premium pay, fringe benefits, as well as
premium
contributions to the State Insurance Fund, social security, medicare
and
private welfare and retirement plans.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Sec. 7. <em>Exemption
of Distressed employers.</em> - Distressed employers shall qualify for
exemption
from the requirement of the <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
upon prior authorization by the Secretary of Labor. Petitions for
exemptions
may be filed within the nearest regional office having jurisdiction
over
the employer not later than January 15, 1976. The regional offices
shall
transmit the petitions to the Secretary of Labor within 24 hours from
receipt
thereof.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Sec. 8. <em>Report
of compliance. </em>- Every covered employer shall make a report of his
compliance with the <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
to the nearest regional labor office not later than January 15 of each
year.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">The report
shall conform substantially with the following form:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</strong>
<center>
<strong>
<font face="Verdana">
</font>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#524744">REPORT
ON COMPLIANCE WITH <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">P.D.
NO. 851</a>
</font>
</font>
</strong>
</center>
</div>
<ul style="text-align: justify;">
<ul>
<strong>
<font face="Verdana">
<font color="#524744">1. Name of
establishment</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">2. Address</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">3.
Principal
product or business</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">4. Total
employment</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">5. Total
number
of workers benefited</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">6. Amount
granted
per employee</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">7. Total
amount
of benefits granted</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">8. Name,
position
and tel. no. of person giving information</font>
</font>
</strong>
</ul>
</ul>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">Sec. 9. <em>Adjudication
of claims. </em>- Non-payment of the thirteenth-month pay provided by
the
<a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
and these rules shall be treated as money claims cases and shall be
processed
in accordance with the Rules Implementing the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code of the Philippines</a> and the Rules of the National Labor
Relations
Commission.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Sec. 10.<em>
Prohibition against reduction or elimination of</em>
</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">
<em>benefits.</em>
- Nothing herein shall be construed to authorize any employer to
eliminate,
or diminish in any way, supplements, or other employee benefits or
favorable
practice being enjoyed by the employee at the time of promulgation of
this
issuance.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Sec. 11. <em>Transitory
Provision.</em> - These rules and regulations shall take effect
immediately
and for purposes of the 13th-month pay for 1975, the same shall apply
only
to those who are employees as of December 16, 1975.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
</div>
<div align="right">
<strong>
<font face="Verdana">
<font color="#524744">Manila,
Philippines, 22 December 1975.</font>
</font>
</strong>
</div>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<center>
<font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="#CHAN%20ROBLES%20AND%20ASSOCIATES%20LAW%20FIRM%20-%20PHILIPPINE%20LABOR%20CIRCULAR-ON-LINE">Back
to Top</a>
</font>
</font>
</font>&nbsp;&nbsp; -&nbsp; <font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="index.htm">Back
to Home</a>
</font>
</font>
</font>&nbsp;&nbsp; -&nbsp; <font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="legal4.htm">Back
to Philippine Labor Circular Index</a>
</font>
</font>
</font>
</center>
&nbsp;<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtualawlibrary</span><br>
<br>
<center>
<strong>
<a name="CHAN ROBLES &amp; ASSOCIATES LAW FIRM-BOTTOM">
</a>
</strong>
</center>
<center>
<strong>
<tt>
<font color="#333333">
<font size="-2">Since
19.07.98</font></font></tt>
</strong>
</center>
<br>
</center>"""),
test.collect_content_individually_bs("""<center>
&nbsp;
<center>
<a name="PRESIDENTIAL DECREE NO. 851-SUPPLEMENTARY RULES">
</a>
</center>
&nbsp;
<center>
<strong>
<tt>
<font color="#993300">
<font size="+1">PRESIDENTIAL
DECREE
NO. 851</font>
</font>
</tt>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#ff0000">&nbsp;</font>
</font>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#524744">SUPPLEMENTARY
RULES AND REGULATIONS</font>
</font>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#524744">IMPLEMENTING
P.D. NO. 851<br>
&nbsp;&nbsp; <br>
</font>
</font>
</strong>
</center>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">To insure uniformity
in the interpretation, application and enforcement of the provisions of
<a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Presidential
Decree No. 851</a> and its <a href="https://www.chanrobles.com/presidentialdecreeno851rules.htm">implementing
regulations</a>, the following clarifications are hereby made for the
information
and guidance of all concerned:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">1. Contractors
and Subcontractors, including Security and Watchman Agencies, are
exempt
for the year 1975 subject to the following conditions:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</strong>
</div>
<ul style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">(a) that the
contracts of such enterprises were entered into before December 16,
1975;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(b) that such
enterprises have complied with all labor standards laws during the year;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(c) that the
contract cannot really accomodate 13-month pay or its equivalent; and</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(d) that the
contract does not provide for cost escalation clause.</font>
</font>
</strong>
</ul>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">This exemption
is without prejudice on the part of the workers to negotiate with their
employers or to seek payment thereof by filing appropriate complaints
with
the Regional Offices of the Department of Labor.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">2. Private
school teachers, including faculty members of colleges and
universities,
are entitled to 1/12 of their annual basic pay regardless of the number
of months they teach or are paid within a year.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">3. New
establishments
operating for less than one year are not covered except subsidiaries or
branches of foreign and domestic corporations.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">4. Overtime
pay, earnings and other remunerations which are not part of the basic
salary
shall not be included in the computation of the 13th-month pay.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">5. In view
of the lack of sufficient time for the dissemination of the provisions
of <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">P.D.
No. 851</a> and its <a href="https://www.chanrobles.com/presidentialdecreeno851rules.htm">Rules</a>
and the unavailability of adequate cash flow due to the long holiday
season,
compliance and reporting of compliance with this Decree are hereby
extended
up to March 31, 1976 except in private schools where compliance for
1975
may be made not later than 30 June 1976.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">6. Nothing
herein shall sanction the withdrawal or diminution of any compensation,
benefits or any supplements being enjoyed by the employees on the
effective
date of this issuance.</font>
</font>
</strong>
<br>
</div>
<strong>
<font face="Verdana">
</font>
</strong>
<div align="right">
<strong>
<font face="Verdana">
<font color="#524744">Manila,
January 16, 1976</font>
</font>
</strong>
</div>
<strong>
<font face="Verdana">
</font>
</strong>
<center>
<font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="#CHAN%20ROBLES%20AND%20ASSOCIATES%20LAW%20FIRM%20-%20PHILIPPINE%20LABOR%20CIRCULAR-ON-LINE">Back
to Top</a>
</font>
</font>
</font>&nbsp;&nbsp; -&nbsp; <font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="index.htm">Back
to Home</a>
</font>
</font>
</font>&nbsp;&nbsp; -&nbsp; <font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="legal4.htm">Back
to Philippine Labor Circular Index</a>
</font>
</font>
</font>
</center>
&nbsp;
<br>
<div style="text-align: center;">&nbsp;<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtualawlibrary</span><br>
<br>
</div>
<br>
<center>
<strong>
<a name="CHAN ROBLES &amp; ASSOCIATES LAW FIRM-BOTTOM">
</a>
</strong>
</center>
<center>
<strong>
<tt>
<font color="#333333">
<font size="-2">Since
19.07.98</font></font></tt>
</strong>
</center>
<br>
</center>"""),
test.collect_content_individually_bs("""<center>
&nbsp;
<center>
<a name="ADMINISTRATIVE ORDER NO. 2, SERIES OF 1976">
</a>
</center>
&nbsp;
<center>
<strong>
<tt>
<font color="#993300">
<font size="+1">PRESIDENTIAL
DECREE
NO. 851</font>
</font>
</tt>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#524744">ADMINISTRATIVE
ORDER NO. 2,</font>
</font>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#524744">SERIES
OF 1976<br>
&nbsp;&nbsp; <br>
</font>
</font>
</strong>
</center>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">In the interest
of public service and efficiency, more particularly to facilitate the
disposition
of cases involving petitions for exemption, complaints, enforcement and
implementation of <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">P.D.
No. 851</a> and its <a href="https://www.chanrobles.com/presidentialdecreeno851rules.htm">implementing
rules and regulations</a>, the following guidelines shall be followed:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<center>
<strong>
<font face="Verdana">
<font color="#cc6600">I.
Petition
for exemption</font>
</font>
</strong>
</center>
<strong>
<font face="Verdana">
<font color="#cc6600">&nbsp;&nbsp;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">1. The Regional
Office concerned shall transmit immediately the petition for exemption
to the Chairman, Wage Commission with comments and recommendations, if
any.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">The petition
shall contain a sworn statement on the inability to implement the <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">Decree</a>
and the reasons, therefore, and shall be accompanied by the following
documents
and statements:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</strong>
</div>
<ul style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">(a) A
certified
true copy of the income tax returns for the last two (2) years;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(b) A
certified
copy of the financial reports for the last two (2) years filed with the
Government entities, such as the Securities and Exchange Commission,
Department
of Trade, Department of Industries and Board of Investments;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(c) A
detailed
sworn statement of the actual monthly losses not covered by the report
required under paragraph (b) above and such other proofs or documents
as
may be required by the Chairman, Wage Commission to establish such
exemption.</font>
</font>
</strong>
</ul>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">2. The Chairman,
Wage Commission and the duly designated staff, shall evaluate all
petitions
for exemption and make appropriate recommendations within 20 working
days
from receipt of the petition to the Secretary of Labor.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">3. Whenever
a petition for exemption has been filed, and complaint for</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">non-compliance
shall be held in abeyance pending the disposition or resolution of the
petition for exemption.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<center>
<strong>
<font face="Verdana">
<font color="#cc6600">II.
Complaint,
enforcement and/or implementation</font>
</font>
</strong>
</center>
<center>
<strong>
<font face="Verdana">
<font color="#cc6600">&nbsp;</font>
</font>
</strong>
</center>
<strong>
<font face="Verdana">
<font color="#524744">1. All
complaints
for non-payment of the 13th-month pay shall be filed with the Field
Services
Division of the Regional Office concerned. The Regional Director shall
direct the said Division to conduct an inspection and investigation in
connection with the complaint filed.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">2. The Field
Service Division of the Regional Office concerned shall see to it that
all covered employees comply with <a href="https://www.chanrobles.com/presidentialdecreeno851.htm">P.D.
No. 851</a>.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">The Regional
Director shall submit a monthly progress report of compliance with the
Decree.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">The reports
of the Regional Offices shall be submitted to the LSS and BLS, and
shall
contain the following:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</strong>
</div>
<ul style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">(a) The total
number of establishments;</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(b) total
number
of workers benefited; and</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">(c) total
amount
of benefits paid.</font>
</font>
</strong>
</ul>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">3. The Regional
Office shall compile, analyze and evaluate compliance reports and
update
the listing of establishments on the basis of the reports submitted.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">Any prior
order,
circular, instruction or memorandum or parts thereof, inconsistent
herewith
are hereby revoked.</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">This Order
shall take effect immediately.</font>
</font>
</strong>
<br>
</div>
<strong>
<font face="Verdana">
</font>
</strong>
<div align="right">
<strong>
<font face="Verdana">
<font color="#524744">Manila,
9 January 1976.&nbsp;</font>
</font>
</strong>
</div>
<strong>
<font face="Verdana">
</font>
</strong>
<br>
<strong>
<font face="Verdana">
</font>
</strong>
<center>
<font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="#CHAN%20ROBLES%20AND%20ASSOCIATES%20LAW%20FIRM%20-%20PHILIPPINE%20LABOR%20CIRCULAR-ON-LINE">Back
to Top</a>
</font>
</font>
</font>&nbsp;&nbsp; -&nbsp; <font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="index.htm">Back
to Home</a>
</font>
</font>
</font>&nbsp;&nbsp; -&nbsp; <font face="Verdana">
<font color="#333333"> <font size="-2">
<a href="legal4.htm">Back
to Philippine Labor Circular Index</a>
</font>
</font>
</font>
</center>
&nbsp;<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtualawlibrary</span><br>
<br>
<center>
<strong>
<a name="CHAN ROBLES &amp; ASSOCIATES LAW FIRM-BOTTOM">
</a>
</strong>
</center>
<center>
<strong>
<tt>
<font color="#333333">
<font size="-2">Since
19.07.98</font></font></tt>
</strong>
</center>
<br>
</center>"""),
test.collect_content_individually_bs("""<div class="Section1">
<center>
<strong>
<font face="Verdana">
<font color="#333333">
<font size="-0">REVISED
GUIDELINES <br>
ON THE IMPLEMENTATION OF THE 13TH MONTH PAY LAW.</font>
</font>
</font>
</strong>
</center>
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
</font>
</font>
</font>
</strong>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
1.&nbsp;&nbsp;Removal
of Salary Ceiling.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
On
August 13, 1986, President Corazon C. Aquino issued Memorandum Order
No.
28 which provides as follows:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span>
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
"Section
1 of Presidential Decree No. 851 is hereby modified to the extent that
all employers are hereby required to pay all their rank-and-file
employees
a 13th month pay not later than December 24 of every year."
</font>
</font>
</font>
</strong>
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
<span style="color: #dadad2;">chan robles&nbsp; virtual law
library</span>
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
Before
its modification by the aforecited Memorandum Order, P.D. No. 851
excludes
from entitlement to the 13th month pay those employees who were
receiving
a basic salary of more than P1,000.00 a month. With the removal of the
salary ceiling of P1,000.00, all rank and file employees are now
entitled
to a 13th month pay regardless of the amount of basic salary that they
receive in a month if their employers are not otherwise exempted from
the
application of P.D. No. 851. Such employees are entitled to the benefit
regardless of their designation or employment status, and irrespective
of the method by which their wages are paid, provided that they have
worked
for at least one (1) month during a calendar year.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
2.&nbsp;&nbsp;Exempted
Employers.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
The
following employers are still not covered by P.D. No. 851:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span>
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
a.&nbsp;&nbsp;The
Government and any of its political subdivisions, including
government-owned
and controlled corporations, excepts those corporations operating
essentially
as private subsidiaries of the Government;
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
b.&nbsp;&nbsp;Employers
already paying their employees a 13th month pay or more in a calendar
year
or its equivalent at the time of this issuance;
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
c.&nbsp;&nbsp;Employers
of household helpers and persons in the personal service of another in
relation to such workers; and
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
d.&nbsp;&nbsp;Employers
of those who are paid on purely commission, boundary, or task basis,
and
those who are paid a fixed amount for performing specific work,
irrespective
of the time consumed in the performance thereof, except where the
workers
are paid on piece-rate basis in which case the employer shall grant the
required 13th month pay to such workers.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
As
used herein, workers paid on piece-rate basis shall refer to those who
are paid a standard amount for every piece or unit of work produced
that
is more or less regularly replicated, without regard to the time spent
in producing the same.&nbsp;
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
The
term "its equivalent" as used on paragraph (b) hereof shall include
Christmas
bonus, mid-year bonus, cash bonuses and other payments amounting to not
less than 1/12 of the basic salary but shall not include cash and stock
dividends, cost of living allowances and all other allowances regularly
enjoyed by the employee, as well as non-monetary benefits. Where an
employer
pays less than required 1/12th of the employees basic salary, the
employer
shall pay the difference.
</font>
</font>
</font>
</strong>
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
<span style="color: #dadad2;">chan robles&nbsp; virtual law
library</span>
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
3.&nbsp;&nbsp;Who
are Rank-and File Employees.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
The
Labor Code distinguishes a rank-and-file employee from a managerial
employee.
It provides that a managerial employee is one who is vested with powers
of prerogatives to lay down and execute management policies and/or to
hire,
transfer, suspend, lay-off, recall discharge, assign or discipline
employees,
or to effectively recommend such managerial actions. All employees not
falling within this definition are considered rank-and-file employees.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
The
above distinction shall be used as guide for the purpose of determining
who are rank-and-file employees entitled to the mandated 13th month pay.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
4.&nbsp;&nbsp;Amount
and payment of 13th Month Pay
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
(a)&nbsp;&nbsp;Minimum
of the Amount. — The minimum 13th month pay required by law shall not
be
less than one-twelfth of the total basic salary earned by an employee
within
a calendar year. For the year 1987, the computation of the 13th month
pay
shall include the cost of living allowances (COLA) integrated into the
basic salary of a covered employee pursuant to Executive Order 178.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
E.O.
No. 178 provides, among other things, that the P9.00 of the daily COLA
of P17.00 for non-agricultural workers shall be integrated into the
basic
pay of covered employees effective 1 May 1987, and the remaining P8.00
effective 1 October 1987. For establishments with less than 30
employees
and paid-up capital of P500,000 or less, the integration of COLAs shall
be as follows: P4.50 effective on 1 May 1987; P4.50 on 1 October 1987;
and P8.00 effective 1 January 1988. Thus, in the computation of the
13th
month pay for 1987, the COLAs integrated into the basic pay shall be
included
as of the date of their integration.&nbsp;
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
Where
the total P17.00 daily COLA was integrated effective 1 May 1987 or
earlier
the inclusion of said COLA as part of the of the basic pay for the
purpose
of computing the 13th month pay shall be reckoned from the date of
actual
integration.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
The
"basic salary" of an employee for the purpose of computing the 13th
month
pay shall include all remunerations or earning paid by this employer
for
services rendered but does not include allowances and monetary benefits
which are not considered or integrated as part of the regular or basic
salary, such as the cash equivalent of unused vacation and sick leave
credits,
overtime, premium, night differential and holiday pay, and
cost-of-living
allowances. However, these salary-related benefits should be included
as
part of the basic salary in the computation of the 13th month pay if by
individual or collective agreement, company practice or policy, the
same
are treated as part of the basic salary of the employees.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
(b)&nbsp;&nbsp;Time
of Payment. — The required 13th month pay shall be paid not later than
December 24 of each year. An employer, however, may give to his
employees
one half (½) of the required 13th month pay before the opening
of
the regular school year and the other half on before the 24th of
December
of every year. The frequency of payment of this monetary benefit may be
the subject of agreement between the employer and the
recognized/collective
bargaining agent of the employees.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
5.&nbsp;&nbsp;13th
Month Pay for Certain Types of Employees.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
(a)&nbsp;&nbsp;Employees
Paid by Results. — Employees who are paid on piece work basis are by
law
entitled to the 13th month pay.&nbsp;
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
Employees
who are paid a fixed or guaranteed wage plus commission are also
entitled
to the mandated 13th month pay, based on their total earnings during
the
calendar year, i.e., on both their fixed or guaranteed wage and
commission.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
(b)&nbsp;&nbsp;Those
with Multiple Employers. — Government employees working part time in a
private enterprise, including private educational institutions, as well
as employees working in two or more private firms, whether on full or
part
time basis, are entitled to the required 13th month pay from all their
private employers regardless of their total earnings from each or all
their
employers.
</font>
</font>
</font>
</strong>
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
<span style="color: #dadad2;">chan robles&nbsp; virtual law
library</span>
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
(c)&nbsp;&nbsp;Private
School Teachers. — Private school teachers, including faculty members
of
universities and colleges, are entitled to the required 13th month pay,
regardless of the number of months they teach or are paid within a
year,
if they have rendered service for at least one (1) month within a year.
</font>
</font>
</font>
</strong>
</p>
<p align="justify"> 
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
6.&nbsp;&nbsp;13th
Month Pay of Resigned or Separated Employee.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
An
employee who has resigned or whose services were terminated at any time
before the time for payment of the 13th month pay is entitled to this
monetary
benefit in proportion to the length of time he worked during the year,
reckoned from the time he started working during the calendar year up
to
the time of his resignation or termination from the service. Thus, if
he
worked only from January up to September his proportionate 13th month
pay
should be equivalent of 1/12 his total basic salary he earned during
that
period.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
The
payment of the 13th month pay may be demanded by the employee upon the
cessation of employer-employee relationship. This is consistent with
the
principle of equity that as the employer can require the employee to
clear
himself of all liabilities and property accountability, so can the
employee
demand the payment of all benefits due him upon the termination of the
relationship.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
7.&nbsp;&nbsp;Non-inclusion
in Regular Wage.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
The
mandated 13th month pay need not be credited as part of regular wage of
employees for purposes of determining overtime and premium pays, fringe
benefits insurance fund, Social Security, Medicare and private
retirement
plans.
</font>
</font>
</font>
</strong>
</p>
<p align="justify">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
8.&nbsp;&nbsp;Prohibitions
against reduction or elimination of benefits.
<span style="color: #dadad2;">chan robles&nbsp; virtual law
library</span>
</font>
</font>
</font>
</strong>
</p>
<div style="text-align: justify;">
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
Nothing
herein shall be construed to authorize any employer to eliminate, or
diminish
in any way, supplements, or other employee benefits or favorable
practice
being enjoyed by the employee at the time of promulgation of this
issuance.
</font>
</font>
</font>
</strong>
<br>
<br>
</div>
<p class="MsoNormal">
<br>
</p>
<center>
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
(Sgd.)
FRANKLIN M. DRILON
</font>
</font>
</font>
</strong>
<br>
<strong>
<font face="Verdana">
<font color="#524744">
<font size="-1">
Secretary
</font>
</font>
</font>
</strong>
</center>
</div>""")
]
```
