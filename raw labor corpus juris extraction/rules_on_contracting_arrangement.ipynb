```python
page_text_content = [
    test.collect_content_individually_bs("""<div style="text-align: justify;"><strong>
        <font face="Verdana"><font color="#524744">By virtue of the
        power vested in the Secretary of Labor and Employment under Articles 5
        (Rule-making) and 106 (Contractor or Subcontractor) of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code of the Philippines, as amended</a>, the following regulations
        governing
        contracting and subcontracting arrangements are hereby issued:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">Section
        1. <em>Guiding principles.</em> - Contracting and subcontracting
        arrangements
        are expressly allowed by law and are subject to regulation for the
        promotion
        of employment and the observance of the rights of workers to just and
        humane
        conditions of work, security of tenure, self-organization, and
        collective
        bargaining. Labor-only contracting as defined herein shall be
        prohibited.</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section 2<em>Coverage.</em>
        - These Rules shall apply to all parties of contracting and
        subcontracting
        arrangements where employer-employee relationship exists. Placement
        activities
        through private recruitment and placement agencies as governed by
        Articles
        25 to 39 of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code</a> are not covered by these Rules.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section 3.
        <em>Trilateral
        Relationship in Contracting Arrangements. -</em> In legitimate
        contracting,
        there exists a trilateral relationship under which there is a contract
        for a specific job, work or service between the principal and the
        contractor
        or subcontractor, and a contract of employment between the contractor
        or
        subcontractor and its workers. Hence, there are three parties involved
        in these arrangements, the principal which decides to farm out a job or
        service to a contractor or subcontractor, the contractor or
        subcontractor
        which has the capacity to independently undertake the performance of
        the
        job, work or service, and the contractual workers engaged by the
        contractor
        or subcontractor to accomplish the job work or service.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section 4.
        <em>Definition
        of Basic Terms.</em> - The following terms as used in these Rules, shall
        mean:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        </p>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        <em>"Contracting"
        </em>or<em> "subcontracting" </em>refers to an arrangement whereby a
        principal
        agrees to put out or farm out with a contractor or subcontractor the
        performance
        or completion of a specific job, work or service within a definite or
        predetermined
        period, regardless of whether such job, work or service is to be
        performed
        or completed within or outside the premises of the principal.</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b) <em>"Contractor
        or subcontractor"</em> refers to any person or entity engaged in a
        legitimate
        contracting or subcontracting arrangement.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(c) <em>"Contractual
        employee"</em> includes one employed by a contractor or subcontractor to
        perform or complete a job, work or service pursuant to an arrangement
        between
        the latter and a principal.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(d) <em>"Principal"</em>
        refers to any employer who puts out or farms out a job, service or work
        to a contractor or subcontractor.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">Section 5. <em>Prohibition
        against labor-only contracting. -</em> Labor-only contracting is hereby
        declared prohibited. For this purpose, labor-only contracting shall
        refer
        to an arrangement where the contractor or subcontractor merely
        recruits,
        supplies or places workers to perform a job, work or service for a
        principal,
        and any of the following elements are present:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(i)
        The contractor or subcontractor does not have substantial capital or
        investment
        which relates to the job, work or service to be performed and the
        employees
        recruited, supplied or placed by such contractor or subcontractor are
        performing
        activities which are directly related to the main business of the
        principal;
        or</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(ii) the
        contractor
        does not exercise the right to control over the performance of the work
        of the contractual employee.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">The foregoing
        provisions
        shall be without prejudice to the application of Article 248 (C ) of
        the
        <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code, as amended</a>.</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">
        <em>"Substantial
        capital or investment"</em> refers to capital stocks and subscribed
        capitalization
        in the case of corporations, tools, equipment, implements, machineries
        and work premises, actually and directly used by the contractor or
        subcontractor
        in the performance or completion of the job, work or service contracted
        out.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">The <em>"right
        to control"</em> shall refer to the right reserved to the person for
        whom
        the services of the contractual workers are performed, to determine not
        only the end to be achieved, but also the manner and means to be used
        in
        reaching that end.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section 6.
        <em>Prohibitions</em>.
        - Notwithstanding Section 5 of these Rules, the following are hereby
        declared
        prohibited for being contrary to law or public policy:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        </p>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        Contracting out of a job, work or service when not done in good faith
        and
        not justified by the exigencies of the business and the same results in
        the termination of regular employees and reduction of work hours or
        reduction
        or splitting of the bargaining unit;</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b)
        Contracting
        out of work with a <em>"cabo"</em> as defined in Section 1 (ii), Rule I,
        Book V of these Rules. <em>"Cabo"</em> refers to a person or group of
        persons
        or to a labor group which, in the guise of a labor organization,
        supplies
        workers to an employer, with or without any monetary or other
        consideration
        whether in the capacity of an agent of the employer or as an ostensible
        independent contractor;</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(c)
        Taking undue
        advantage of the economic situation or lack of bargaining strength of
        the
        contractual employee, or undermining his security of tenure or basic
        rights,
        or circumventing the provisions of regular employment, in any of the
        following
        instances:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        </p>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(i)
        In addition to his assigned functions, requiring the contractual
        employee
        to perform functions which are currently being performed by the regular
        employees of the principal or of the contractor or subcontractor;</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(ii)
        Requiring
        him to sign, as a precondition to employment or continued employment,
        an
        antedated resignation letter; a blank payroll; a waiver of labor
        standards
        including minimum wages and social or welfare benefits; or a quitclaim
        releasing the principal, contractor or subcontractor from any liability
        as to payment of future claims; and</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(iii)
        Requiring
        him to sign a contract fixing the period of employment to a term
        shorter
        than the term of the contract between the principal and the contractor
        or subcontractor, unless the latter contract is divisible into phases
        for
        which substantially different skills are required and this is made
        known
        to the employee at the time of engagement;</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">(d)
        Contracting
        out of a job, work or service through an in-house agency which refers
        to
        a contractor or subcontractor engaged in the supply of labor which is
        owned,
        managed or controlled by the principal and which operates solely for
        the
        principal;</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(e)
        Contracting
        out of a job, work or service directly related to the business or
        operation
        of the principal by reason of a strike or lockout whether actual or
        imminent;</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(f)
        Contracting
        out of a job, work or service being performed by union members when
        such
        will interfere with, restrain or coerce employees in the exercise of
        their
        rights to self organization as provided in Art. 248 (c) of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code, as amended</a>.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">Section 7. <em>Existence
        of an employer-employee relationship</em>. - The contractor or
        subcontractor
        shall be considered the employer of the contractual employee for
        purposes
        of enforcing the provisions of the Labor Code and other social
        legislation.
        The principal, however, shall be solidarily liable with the contractor
        in the event of any violation of any provision of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code</a>, including the failure to pay wages.</font></font></strong>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">The
        principal shall be deemed the employer of the contractual employee in
        any
        of the following cases as declared by a competent authority:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(a)
        where there
        is labor-only contracting; or</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b)
        where the
        contracting arrangement falls within the prohibitions provided in
        Section
        6 (Prohibitions) hereof.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">Section 8. <em>Rights
        of Contractual Employees</em>. - Consistent with Section 7 of these
        Rules,
        the contractual employee shall be entitled to all the rights and
        privileges
        due a regular employee as provided for in the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code, as amended</a>, to include the following:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        Safe and healthful working conditions;</font></font></strong>
        <br><strong>
        <font face="Verdana"><font color="#524744">(b) Labor
        standards
        such as service incentive leave, rest days, overtime pay, holiday pay,
        13th month pay and separation pay;</font></font></strong>
        <br><strong>
        <font face="Verdana"><font color="#524744">(c) Social
        security and welfare benefits;</font></font></strong>
        <br><strong>
        <font face="Verdana"><font color="#524744">(d)
        Self-organization,
        collective bargaining and peaceful concerted action; and</font></font></strong>
        <br><strong>
        <font face="Verdana"><font color="#524744">(e)
        Security
        of tenure.</font></font></strong>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">Section 9. <em>Contract
        between contractor or subcontractor and contractual employee</em>. -
        Notwithstanding
        oral or written stipulations to the contrary, the contract between the
        contractor or subcontractor and the contractual employee, which shall
        be
        in writing, shall include the following terms and conditions:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        The specific description of the job, work or service to be performed by
        the contractual employee;</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b) The
        place
        of work and terms and conditions of employment, including a statement
        of
        the wage rate applicable to the individual contractual employee; and</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(c) The
        term
        or duration of employment, which shall be coextensive with the contract
        of the principal and subcontractor, or with the specific phase for
        which
        the contractual employee is engaged, as the case may be.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">The
        contractor
        or subcontractor shall inform the contractual employee of the foregoing
        terms and conditions on or before the first day of his employment.</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        10.
        <em>Effect
        of Termination of Contractual Employment. -</em> In cases of termination
        of employment prior to the expiration of the contract between the
        principal
        and the contractor or subcontractor, the right of the contractual
        employee
        to separation pay or other related benefits shall be governed by the
        applicable
        laws and jurisprudence on termination of employment.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Where the
        termination
        results from the expiration of the contract between the principal and
        the
        contractor or subcontractor, or from the completion of the phase of the
        job, work or service for which the contractual employee is engaged, the
        latter shall not be entitled to separation pay. However, this shall be
        without prejudice to completion bonuses or other emoluments, including
        retirement pay as may be provided by law or in the contract between the
        principal and the contractor or subcontractor.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        11.
        <em>Registration
        of Contractors or Subcontractors</em>. - Consistent with the authority
        of
        the Secretary of Labor and Employment to restrict or prohibit the
        contracting
        out of labor through appropriate regulations, a registration system to
        govern contracting arrangements and to be implemented by the Regional
        Offices
        is hereby established.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">The
        registration
        of contractors and subcontractors shall be necessary for purposes of
        establishing
        an effective labor market information and monitoring.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Failure to
        register
        shall give rise to the presumption that the contractor is engaged in
        labor-only
        contracting.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        12.
        <em>Requirements
        for registration</em>. - A contractor or subcontractor shall be listed
        in
        the registry of contractors and subcontractors upon completion of an
        application
        form to be provided by the DOLE. The applicant contractor or
        subcontractor
        shall provide in the application form the following information:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        </p>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        The name and business address of the applicant and the area or areas
        where
        it seeks to operate;</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b) The
        names
        and addresses of officers, if the applicant is a corporation,
        partnership,
        cooperative or union;</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(c) The
        nature
        of the applicant's business and the industry or industries where the
        applicant
        seeks to operate;</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(d) The
        number
        of regular workers; the list of clients, if any; the number of
        personnel
        assigned to each client, if any and the services provided to the client;</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(e) The
        description
        of the phases of the contract and the number of employees covered in
        each
        phase, where appropriate; and</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(f) A
        copy of
        audited financial statements if the applicant is a corporation,
        partnership,
        cooperative or a union, or copy of the latest ITR if the applicant is a
        sole proprietorship.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">The
        application
        shall be supported by:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        A certified copy of a certificate of registration of firm or business
        name
        from the Securities and Exchange Commission (SEC), Department of Trade
        and Industry (DTI), Cooperative Development Authority (CDA), or from
        the
        DOLE if the applicant is a union; and</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b) A
        certified
        copy of the license or business permit issued by the local government
        unit
        or units where the contractor or subcontractor operates.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">The
        application
        shall be verified and shall include an undertaking that the contractor
        or subcontractor shall abide by all applicable labor laws and
        regulations.</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        13.
        <em>Filing
        and processing of applications</em>. - The application and its
        supporting
        documents shall be filed in triplicate in the Regional Offices where
        the
        applicant principally operates. No application for registration shall
        be
        accepted unless all the foregoing requirements are complied with. The
        contractor
        or subcontractor shall be deemed registered upon payment of a
        registration
        fee of P100.00 to the Regional Office.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Where all
        the
        supporting documents have been submitted, the Regional Office shall
        deny
        or approve the application within seven (7) working days after its
        filing.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Upon
        registration,
        the Regional Office shall return one set of the duly-stamped
        application
        documents to the applicant, retain one set for its file, and transmit
        the
        remaining set to the Bureau of Local Employment. The Bureau shall
        devise
        the necessary forms for the expeditious processing of all applications
        for registration.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        14.
        <em>Duty
        to produce copy of contract between the principal and the contractor or
        subcontractor. -</em> The principal or the contractor or subcontractor
        shall
        be under an obligation to produce a copy of the contract between the
        principal
        and the contractor in the ordinary course of inspection. The contractor
        shall likewise be under an obligation to produce a copy of the contract
        of employment of the contractual worker when directed to do so by the
        Regional
        Director or his authorized representative.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">A copy of
        the
        contract between the contractual employee and the contractor or
        subcontractor
        shall be furnished the certified bargaining agent, if there is any.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        15.
        <em>Annual
        Reporting of Registered Contractors</em>. - The contractor or
        subcontractor
        shall submit in triplicate its annual report using a prescribed form to
        the appropriate Regional Office not later than the 15th of January of
        the
        following year. The report shall include:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        </p>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        A list of contracts entered with the principal during the subject
        reporting
        period;</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b) The
        number
        of workers covered by each contract with the principal;</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(c) A
        sworn
        undertaking that the benefits from the Social Security System (SSS),
        the
        Home Development Mutual Fund (HDMF), PhilHealth, Employees Compensation
        Commission (ECC), and remittances to the Bureau of Internal Revenue
        (BIR)
        due its contractual employees have been made during the subject
        reporting
        period.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">The Regional
        Office
        shall return one set of the duly-stamped report to the contractor or
        subcontractor,
        retain one set for its file, and transmit the remaining set to the
        Bureau
        of Local Employment within five (5) days from receipt thereof.</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        16.
        <em>Delisting
        of contractors or subcontractors</em>. - Subject to due process, the
        Regional
        Director shall cancel the registration of contractors or subcontractors
        based on any of the following grounds:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        </p>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        Non-submission of contracts between the principal and the contractor or
        subcontractor when required to do so;</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b)
        Non-submission
        of annual report;</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(c)
        Findings
        through arbitration that the contractor or subcontractor has engaged in
        labor-only contracting and the prohibited activities as provided in
        Section
        6 (Prohibitions) hereof; and</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">(d)
        Non-compliance
        with labor standards and working conditions.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">Section 17. <em>Renewal
        of registration of contractors or subcontractors</em>. - All registered
        contractors or subcontractors may apply for renewal of registration
        every
        three years. For this purpose, the Tripartite Industrial Peace Council
        (TIPC) as created under Executive Order No. 49, shall serve as the
        oversight
        committee to verify and monitor the following:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
        <blockquote><strong>
        <font face="Verdana"><font color="#524744">(a)
        Engaging in allowable contracting activities; and</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">(b)
        Compliance
        with administrative reporting requirements.</font></font></strong>
        </p>
        </blockquote><strong>
        <font face="Verdana"><font color="#524744">Section 18. <em>Enforcement
        of Labor Standards and Working Conditions</em>. - Consistent with
        Article
        128 (Visitorial and Enforcement Power) of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code, as amended</a>, the Regional Director through his duly authorized
        representatives, including labor regulation officers shall have the
        authority
        to conduct routine inspection of establishments engaged in contracting
        or subcontracting and shall have access to employer's records and
        premises
        at any time of the day or night whenever work is being undertaken
        therein,
        and the right to copy therefrom, to question any employee and
        investigate
        any fact, condition or matter which may be necessary to determine
        violations
        or which may aid in the enforcement of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code</a> and of any labor law, wage order, or rules and regulations
        issued
        pursuant thereto.</font></font></strong>
        <p><strong>
        <font face="Verdana"><font color="#524744">The
        findings
        of the duly authorized representative shall be referred to the Regional
        Director for appropriate action as provided for in Article 128, and
        shall
        be furnished the collective bargaining agent, if any.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Based on
        the
        visitorial and enforcement power of the Secretary of Labor and
        Employment
        in Article 128 (a), (b), (c) and (d), the Regional Director shall issue
        compliance orders to give effect to the labor standards provisions of
        the
        <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
        Code</a>, other labor legislation and these guidelines.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        19.
        <em>Solidary
        liability</em>. - The principal shall be deemed as the direct employer
        of
        the contractual employees and therefore, solidarily liable with the
        contractor
        or subcontractor for whatever monetary claims the contractual employees
        may have against the former in the case of violations as provided for
        in
        Sections 5 (Labor-Only contracting), 6 (Prohibitions), 8 (Rights of
        Contractual
        Employees) and 16 (Delisting) of these Rules. In addition, the
        principal
        shall also be solidarily liable in case the contract between the
        principal
        and contractor or subcontractor is preterminated for reasons not
        attributable
        to the fault of the contractor or subcontractor.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        20.
        <em>Supersession.</em>
        - All rules and regulations issued by the Secretary of Labor and
        Employment
        inconsistent with the provisions of this Rule are hereby superseded.
        Contracting
        or subcontracting arrangements in the construction industry, under the
        licensing coverage of the PCAB and shall not include shipbuilding and
        ship
        repairing works, however, shall continue to be governed by Department
        Order
        No. 19, series of 1993.</font></font></strong>
        </p>
        <p><strong>
        <font face="Verdana"><font color="#524744">Section
        21.
        <em>Effectivity</em>.
        - This Order shall be effective fifteen (15) days after completion of
        its
        publication in two (2) newspapers of general circulation.</font></font></strong>
        </p>
        </blockquote>
        </div>"""
    ),
    test.collect_content_individually_bs("""<center>
<a name="CHAN ROBLES AND ASSOCIATES LAW FIRM - PHILIPPINE LABOR CIRCULAR-ON-LINE">
</a>
<a name="TOP OF LABOR CODE PAGE">
</a>
<br>
<br>
<strong>
<font face="CG Times"><font color="#8b7b45"><font size="+1">
<img src="" style="width: 520px; height: 66px;"></font>
</font></font></strong>
<br>
<br>
<center><br><strong>
<font face="Verdana"><font color="#330000">This web page
features
the full text of</font></font></strong>
<br><strong>
<font face="Verdana"><font color="#330000">
<a href="#Department%20Order%20No.%2003">Department
Order No. 03 (Series of 2001)</a>.</font></font></strong>
<br><strong>
<font face="Verdana"><font color="#8b7b45"><font size="-1">REVOKING
<a href="https://www.chanrobles.com/doledepartmentorderno10-1997.htm">DEPARTMENT
ORDER NO. 10, SERIES OF 1997</a>, AND CONTINUING TO PROHIBIT LABOR-ONLY
CONTRACTING</font></font>
</font></strong>
<br></center>
<br>
<center>
<center>
<table style="" border="0" cellpadding="5" cellspacing="5" cols="1" width="45%">
<tbody>
<tr>
<td>
<center><strong>
<font face="Verdana"><font color="#524744"><font size="-1">Read
Full Text of the Rules on Contracting and Subcontracting Arrangement:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</font></strong></center>
</td>
</tr>
<tr>
<td>
<center>
<li><strong>
<font color="#524744"><font size="-2">
<a href="https://www.chanrobles.com/doledepartmentorderno18-02-2002.html"><font face="Verdana">Department
Order No. 18-02 (Series of 2002) <em>[</em></font>
<em>
<font face="Verdana">Rules
Implementing Articles 106 to 109 of the Labor Code, as Amended]</font>
</em>
</a></font>
</font></strong>
</li></center>
</td>
</tr>
<tr>
<td>
<center>
<li><strong>
<font face="Verdana"><font color="#524744"><font size="-1">
<a href="https://www.chanrobles.com/doledepartmentorderno03-2001.html">Department
Order No. 03 (Series of 2001) <em>[Revoking Department Order No. 10,
Series
of 1997, and Continuing to Prohibit Labor-Only Contracting]</em>
</a></font>
</font></font></strong>
</li></center>
</td>
</tr>
<tr>
<td>
<center><strong>
<font face="Verdana"><font color="#330000"><font size="-2">
<a href="https://www.chanrobles.com/dolecontractingsubcontractingprimer.html" onmouseover="window.status='DOLE PRIMER ON CONTRACTING AND SUBCONTRACTING'; return true" onmouseout="window.status='';return true">DOLE
Primer on Contracting and Subcontracting Under Articles 106 to 109 of
the
Labor Code</a></font>
</font></font></strong></center>
</td>
</tr>
<tr>
<td>
<center><strong>
<font face="Verdana"><font size="-2">
<a href="https://www.chanrobles.com/doledepartmentorderno10-1997.htm" onmouseover="window.status='DEPARTMENT ORDER NO. 10, SERIES OF 1997'; return true" onmouseout="window.status='';return true"><font color="#330000">Department
Order No. 10 (Series of 1997) [</font>
<font color="#333300">Amending
the
Rules Implementing Books III and VI of the Labor Code, as Amended]</font>
</a></font>
</font></strong></center>
</td>
</tr>
</tbody>
</table></center>
</center>
<center>
<center>
<table border="0" cols="1" height="10%" width="30%">
<tbody>
<tr>
<td>
<center><strong>
<font face="Verdana"><font color="#ff6600"><font size="-2">
<a href="https://www.chanrobles.com/virtualibrary.htm" onmouseover="window.status='CHAN ROBLES VIRTUAL LAW LIBRARY '; return true" onmouseout="window.status='';return true">THE
CHAN ROBLES VIRTUAL LAW LIBRARY - <blink>QUICK</blink> GLANCE</a></font>
</font></font></strong></center>
</td>
</tr>
<tr>
<td>
<center>
<font size="-2"><strong>
<font face="Verdana">
<a href="https://www.chanrobles.com/virtualibrary1.htm" onmouseover="window.status='PHILIPPINE LEGAL RESOURCES - VIRTUAL LAW LIBRARY OF CHAN ROBLES'; return true" onmouseout="window.status='';return true"><font color="#8b7b45">Philippines</font>
</a>
<font color="#0000ff">&nbsp;</font></font></strong>
<font face="Verdana"><font color="#0000ff"></font>
</font>
<font color="#524744">|&nbsp; </font>
<font face="Verdana"><strong>
<font color="#8b7b45">
<a href="https://www.chanrobles.com/legal2ab.htm" onmouseover="window.status='WORLDWIDE LEGAL RESOURCES - VIRTUAL LAW LIBRARY OF CHAN ROBLES'; return true" onmouseout="window.status='';return true">Worldwide</a></font></strong>
<font color="#0000ff">&nbsp;
</font></font>
<font color="#524744">|</font>
<font face="Verdana"><font color="#0000ff">&nbsp;
</font><strong>
<font color="#8b7b45">
<a href="https://www.chanrobles.com/virtualibrary2.htm" onmouseover="window.status='THE BUSINESS PAGE - VIRTUAL LAW LIBRARY OF CHAN ROBLES'; return true" onmouseout="window.status='';return true">The
Business Page</a></font></strong></font>
</font></center>
</td>
</tr>
</tbody>
</table></center>
</center>
<center>
<br>
<a name="Department Order No. 03">
</a>
<br>
<br>
<strong>
<font face="Arial Black"><font color="#003300"><font size="-0">DEPARTMENT
ORDER NO. 03</font></font>
</font></strong>
<br><strong>
<font face="Verdana"><font color="#524744">(Series of 2001)</font></font></strong>
<br>
<br><strong>
<font face="Verdana"><font color="#003300"><font size="-0">REVOKING
<a href="https://www.chanrobles.com/doledepartmentorderno10-1997.htm">DEPARTMENT
ORDER NO. 10, SERIES OF 1997</a>,</font></font>
</font></strong>
<br><strong>
<font face="Verdana"><font color="#003300"><font size="-0">AND
CONTINUING TO PROHIBIT LABOR-ONLY CONTRACTING</font></font>
</font></strong>
<br>
<br></center><strong>
<font face="Verdana"><font color="#524744">Pursuant to
Articles
5 and 106 of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code, as amended</a>, the following are hereby issued:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<p style="text-align: justify;"><strong>
<font face="Verdana"><font color="#524744">Section 1. <em>Revocation
of <a href="https://www.chanrobles.com/doledepartmentorderno10-1997.htm">Department
Order No. 10</a>
</em>
<a href="https://www.chanrobles.com/doledepartmentorderno10-1997.htm">.</a>
- <a href="https://www.chanrobles.com/doledepartmentorderno10-1997.htm">Department
Order No. 10, Series of 1997</a>, otherwise known as the rules
implementing
Article 106 to 109 of Book III of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>, is hereby revoked effective immediately.</font></font></strong>
</p>
<p style="text-align: justify;"><strong>
<font face="Verdana"><font color="#524744">Section 2. <em>Prohibition
against labor-only contracting</em>. - Labor-only contracting is hereby
declared prohibited. There is labor-only contracting where the
contractor
or subcontractor merely recruits, supplies or places workers to perform
a job, work or service for a principal, and the following elements are
present:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<div style="text-align: justify;">
<blockquote><strong>
<font face="Verdana"><font color="#524744">(a)
The contractor or subcontractor does not have substantial capital or
investment
to actually perform the job, work or service under its own account and
responsibility; and</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">(b) The
employees
recruited, supplied or placed by such contractor or subcontractor is
performing
activities, which are directly related to the main business of the
principal.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">Section 3. <em>Non-impairment
of existing contracts; Non-diminution of benefits</em>. Subject to the
provisions
of the Civil Code and existing jurisprudence, nothing herein shall
impair
the rights or diminish the benefits being enjoyed by parties to
existing
contracting or subcontracting arrangements.</font></font></strong>
</div> <br>
<div style="text-align: justify;"><strong>
<font face="Verdana"><font color="#524744">Section 4. <em>Effectivity.</em>&nbsp;
- This Order shall be effective fifteen days after publication in two
newspapers
of general circulation until a new set of guidelines implementing
Articles
106 to 109 of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a> shall have been promulgated. Such new guidelines shall be
formulated
upon prior consultations with all sectors concerned, particularly the
Tripartite
Industrial Peace Council (TIPC) established under Executive Order No.
49,
Series of 1998.</font></font></strong>
<br>
</div>
<p>&nbsp;
</p>
<div align="right"><strong>
<font face="Verdana"><font color="#524744">Manila,
Philippines, 08 May 2001</font></font></strong>
</div>
<center>
<br>
<br><strong>
<font face="Verdana"><font color="#003300">PATRICIA A.
STO. TOMAS</font></font></strong>
<br><strong>
<font face="Verdana"><font color="#003300">Secretary</font></font></strong>
<br>
<br>
<font face="Verdana"><font color="#333333"><font size="-2">
<a href="#CHAN%20ROBLES%20AND%20ASSOCIATES%20LAW%20FIRM%20-%20PHILIPPINE%20LABOR%20CIRCULAR-ON-LINE">Back
to Top</a></font>
</font></font>&nbsp;&nbsp; -&nbsp; <font face="Verdana"><font color="#333333"><font size="-2">
<a href="https://www.chanrobles.com">Back
to Home</a></font>
</font></font>&nbsp;&nbsp; -&nbsp; <font face="Verdana"><font color="#333333"><font size="-2">
<a href="https://www.chanrobles.com/legal4.htm">Back
to Philippine Labor Circular Index</a></font>
</font></font>
<br>
<br>
<center><strong>
<a name="CHAN ROBLES &amp; ASSOCIATES LAW FIRM-BOTTOM">
</a></strong></center>
<center><strong>
<tt>
<font color="#333333"><font size="-2">&nbsp;<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtualawlibrary</span><br>
<br>
&nbsp;&nbsp; <br></font></font></tt></strong></center>
</center>
<p><br>
<br>
</p> <br>
<p></p>
</center>"""),
test.collect_content_individually_bs("""<div style="text-align: justify;"><strong>
<font face="Verdana"><font color="#003300">1. WHAT IS CONTRACTING
AND SUBCONTRACTING?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">There
is contracting or subcontracting when an employer, referred to as the
principal,
farms out the performance of a part of its business to another,
referred
to as the contractor or subcontractor. For the purpose of undertaking
the
principal's business that is farmed out, the contractor or
subcontractor
then employs its own employees.</font></font></strong>
<br>
<br><strong>
<font face="Verdana"><font color="#524744">Contracting
and subcontracting are synonymous under Philippine labor law. The term
that is more commonly used is subcontracting.</font></font></strong>
</blockquote><strong>
<font face="Verdana"><font color="#003300">2. IN THE
EMPLOYMENT
OF WORKERS, IS THERE A DIFFERENCE BETWEEN AN ORDINARY EMPLOYER-EMPLOYEE
RELATIONSHIP AND SUBCONTRACTING?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Yes.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">In an
ordinary
employer-employee relationship, there are only two parties involved -
the
employer and the employee. This relationship is established through a
four-fold
test, under which the employer:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">a.
Directly exercises control and supervision over the employee not only
as
to the results of the work but also as to the means employed to attain
this result;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">b. Has
the power
to select and hire the employee;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">c. Has
the obligation
to pay the employees his or her wages and other benefits.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">The power of
control
is the most important factor in determining the existence of an
employer-employee
relationship. The employer need not actually exercise this power. It is
enough that the employer retains the right to exercise this power. It
is
enough that the employer retains the right to exercise it as it may
deem
necessary or appropriate.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">In
subcontracting,
there are three parties involved:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">a.
The principal which decides to farm out a job or service to a
subcontractor;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">b. The
subcontractor
which has the capacity to independently undertake the performance of
the
job or service; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">c. The
employees
engaged by the subcontractor to accomplish the job or service.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">In
subcontracting,
the four-fold test of employer-employee relationship should be
satisfied
by the subcontractor in relation to the employees it engages to
accomplish
the subcontracted job or service. In such cases, the subcontractor is
also
referred to as independent contractor.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">If the
four-fold
test is satisfied not by the subcontractor but by the principal, the
principal
then becomes the employer of the employees engaged to accomplish the
job
or service. What exists is not subcontracting but a direct
employer-employee
relationship between the principal and the employees.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#003300">3. IS THERE A
DIFFERENCE
BETWEEN A SUBCONTRACTOR AND A PRIVATE RECRUITMENT AND PLACEMENT AGENCY
(PRPA)?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Yes.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">A
subcontractor
directly undertakes a specific job or service for a principal, and for
this purpose, employs its own workers. A PRPA cannot be a
subcontractor.
It simply recruits workers for the purpose of placing them with another
employer so that the workers recruited will not become the PRPA's
employees.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">A
subcontractor
is governed by the laws and rules enumerated under Question # 4 below.
A PRPA is governed by Articles 25 to 39 of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a> and the rules implementing these articles.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">A
subcontractor
does not need authority from the Department of Labor and Employment
(DOLE)
to undertake a subcontracted job or service. A PRPA needs an authority
or license from DOLE to legally undertake a recruitment and placement
activities.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#003300">4. WHAT LAW OR
RULES GOVERN SUBCONTRACTING?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">The
basic law governing subcontracting is the <a href="https://www.chanrobles.com/legal4labor.htm">Labor
Code</a>, particularly Articles 106 to 109. These provisions prescribe
the conditions for regulating subcontracting and the rights and
obligations
of parties to this arrangement. There was also a set of rules
implementing
Articles 106 to 109, known as <a href="https://www.chanrobles.com/doledepartmentorderno10.html">Department
Order No. 10</a>, issued by DOLE in 1997. However, <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a> was revoked by DOLE on 08 May 2001 through another order,
<a href="https://www.chanrobles.com/doledepartmentorderno3.html">D. O.
No.
3, Series of 2001</a>. <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3 took effect on 29 May 2001</a>.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">With the
revocation
of <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a>, the following laws and rules will apply in addition to
Articles
106 to 109 of the <a href="https://www.chanrobles.com/legal4labor.htm">Labor
Code</a>:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">a.
Article 248 (c) which disallows contracting out of services or
functions
being performed by union members when such will interfere with,
restrain
or coerce employees in the exercise of their rights to
self-organization;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">b.
Article 280.
which classifies employees into regular, project or seasonal employees;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">c.
Article 2180
of the <a href="https://www.chanrobles.com/civilcodeofthephilippines.htm">Civil
Code</a>, under which the principal, in a civil suit for damages
instituted
by an injured person, can be held liable for any negligent acts of the
employees of a labor-only contractor;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">d.
Republic
Act No. 5487 and its implementing rules, which regulate the operation
of
security agencies;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">e.
Jurisprudence
interpreting the foregoing laws;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">f. <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a>;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">g. D. O.
No.
19, Series of 1993, for subcontracting arrangements in the construction
industry; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">h.
Contractual
stipulations provided these are not in conflict with <a href="https://www.chanrobles.com/legal4labor.htm">Labor
Code</a> provisions, jurisprudence, and <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. Nos. 3</a> and 19.</font></font></strong>
</p>
</blockquote>
</blockquote><strong>
<font face="Verdana"><font color="#003300">5. ASIDE FROM
REVOKING
<a href="https://www.chanrobles.com/doledepartmentorderno10.html">D. O.
NO. 10</a>, WHAT ARE THE IMPORTANT FEATURES OF <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. NO. 3</a>?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">The
following are the important features of <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a>.</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">a.
It prohibits labor-only contracting;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">b. It
recognizes
the continuing validity of contracts entered into when <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a> was still in force;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">c. It is
a temporary
measure;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">d. It
sets the
process and mechanism, which is through consultations through the
Tripartite
Industrial Peace Council, by which a new set of rules shall be
formulated.</font></font></strong>
</p>
</blockquote>
</blockquote><strong>
<font face="Verdana"><font color="#003300">6. DOES <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. NO. 3</a> RENDER SUBCONTRACTING ILLEGAL?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">No,
provided the requirements for legitimate subcontracting are satisfied
and
the prohibition against labor-only subcontracting is observed.</font></font></strong>
</blockquote><strong>
<font face="Verdana"><font color="#003300">7. WHAT IS
LEGITIMATE
SUBCONTRACTING?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Neither
the Labor Code nor <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> has a definition of legitimate subcontracting.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">However,
while
<a href="https://www.chanrobles.com/doledepartmentorderno3.html">D. O.
No.
3</a> rendered <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a> ineffective, existing jurisprudence still provides
definitive
guidance. In two recent cases decided by the Supreme Court (Vinoy v.
National
Labor Relations Commission, G.R. No. 126586, 02 February 2000, and Lim
v. National Labor Relations Commission, G.R. No. 124630, 19 February
1999),
the definition of legitimate subcontracting in <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a> is favorably cited as follows:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">Contracting
shall be legitimate if the following conditions concur:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">a.
the contractor or subcontractor carries on a distinct and independent
business
and undertakes to perform the job, work or service on its own account
and
under its own responsibility, according to its own manner and method,
and
free from the control and direction of the principal in all matters
connected
with the performance of the work except as to the results thereof;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">b. the
contractor
or subcontractor has substantial capital or investment;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">c. The
agreement
between the principal and the contractor or subcontractor assures the
contractual
employees entitlement to all occupational safety and health standards,
free exercise of the right to self organization, security of tenure,
and
social and welfare benefits.</font></font></strong>
</p>
</blockquote>
</blockquote><strong>
<font face="Verdana"><font color="#333300">8. WHAT IS
SUBSTANTIAL
CAPITAL? IS SUBSTANTIAL CAPITAL SUFFICIENT TO ESTABLISH LEGITIMATE
SUBCONTRACTING?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Substantial
capital refers to such investment, whether in the form of money,
facilities,
tools, equipment, machineries, work premises, or subscribed capital
stock
that would indicate the subcontractor's capacity to undertake the
subcontracted
work or service independently. For example, a subcontractor with a
capital
stock of P1 Million which is fully subscribed and paid for has been
deemed
by the Supreme Court to be a highly capitalized venture which satisfies
the requirement of substantial capital.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">Where a
subcontractor
is highly capitalized, the Supreme Court has held that it need not show
evidence that it has investment in the form of tools, equipment,
machineries,
work premises, among others, to be considered legitimate. However, it
is
still necessary for it to show that it has the capacity to be an
independent
contractor, That is, it can undertake the performance of the contract
according
to its own manner and method, free from the supervision of the
principal
in all matters except as to the results of the work.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#003300">9. IS
LEGITIMATE
SUBCONTRACTING DIFFERENT FROM LABOR-ONLY CONTRACTING? HOW IS LABOR-ONLY
CONTRACTING DEFINED?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Yes,
legitimate subcontracting is different from labor-only contracting
because
the former is allowed and the latter is illegal and prohibited.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">Section 2
of
<a href="https://www.chanrobles.com/doledepartmentorderno3.html">D. O.
No.
3</a> states that there is labor-only contracting where the contractor
or subcontractor merely recruits, supplies or places workers to perform
a job, work or service for a principal, and the following elements are
present:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">a.
The contractor or subcontractor does not have substantial capital or
investment
to actually perform the job, work or service under its own account and
responsibility; and</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">b. The
employees
recruited, supplied or placed by such contractor or subcontractors are
performing activities directly related to the main business of the
principal.</font></font></strong>
</p>
</blockquote>
</blockquote><strong>
<font face="Verdana"><font color="#003300">10. WHAT IS THE
BASIS OF THE STATE IN PROHIBITING LABOR-ONLY CONTRACTING? WHAT IS THE
OBJECTIVE
OF THE PROHIBITION?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">The
bases of the State in prohibiting labor-only contracting are:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">a.
The <a href="https://www.chanrobles.com/philsupremelaw1.htm">Constitution</a>,
which mandates that the State shall protect labor and promote its
welfare,
and shall guarantee basic labor rights including just and humane terms
and conditions of employment and the right to self-organization.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">b.
Article 106
of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>, which allows the Secretary of Labor to distinguish between
labor-only
contracting and job contracting to prevent any violation or
circumvention
of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">The objective
of
the State in prohibiting labor-only contracting is to ensure that labor
laws are followed and to prevent exploitation of workers. A labor-only
contractor is one which presents itself as an employer even if it does
not have capital to run a business or capacity to ensure that its
workers
are paid their wages and other benefits as prescribed by law. As such,
it cannot independently undertake to perform a subcontracted job or
service.
To allow a labor-only contractor to operate is to give it an
opportunity
to circumvent the law and to exploit workers.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">
<a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> is not the first regulation to prohibit labor-only
contracting.
The prohibition was embodied in the original rules implementing
Articles
106 to 109 issued right after the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a> took effect in 1974. <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a> also contained a similar prohibition. <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> merely reiterates the prohibition.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#003300">11. <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. NO. 10</a> ENUMERATED ACTIVITIES PERMITTED FOR SUBCONTRACTING. NOW
THAT
IT HAS BEEN REVOKED, DOES THIS MEAN THAT SUCH ACTIVITIES MAY NO LONGER
BE SUBCONTRACTED?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Not
necessarily. These activities may still be subcontracted provided (a)
the
laws and rules under Question # 4 are observed; and (b) the conditions
for legitimate contracting under Question # 7 and the prohibition
against
labor-only contracting under Question # 9 are met.</font></font></strong>
</blockquote><strong>
<font face="Verdana"><font color="#003300">12. <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. NO. 10</a> ENUMERATED PROHIBITED ACTIVITIES. NOW THAT IT HAS BEEN
REVOKED,
ARE THERE STILL ANY PROHIBITED SUBCONTRACTING ARRANGEMENTS?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Yes.
Expressly prohibited are (a) labor-only contracting as defined in <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a>; and (b) contracting out of services being performed by
union
members when such will interfere with, restrain or coerce employees in
the exercise of their right to self-organization under Article 248 (c)
of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>.</font></font></strong>
</blockquote><strong>
<font face="Verdana"><font color="#003300">13. WHAT WILL
BE
THE EFFECT OF A LABOR-ONLY CONTRACTING ARRANGEMENT?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">The
following are the effects:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">a.
The subcontractor will be treated as the agent of the principal. Since
the act of an agent is the act of the principal, representations made
by
the subcontractor to the employees will bind the principal.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">b. The
principal
will become the employer as if it directly employed the workers engaged
to undertake the subcontracted job or service. It will be responsible
to
them for all their entitlements and benefits under the labor laws.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">c. The
principal
and the subcontractor will be solidarily treated as the employer.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">d. The
employees
will become employees of the principal, subject to the classifications
of employees under Article 28 of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">If the
labor-only
contracting activity is undertaken by a legitimate labor organization,
a petition for cancellation of union registration may be filed against
it, pursuant to Article 239(e).</font></font></strong>
</blockquote><strong>
<font face="Verdana"><font color="#003300">14. IF A
LEGITIMATE
SUBCONTRACTOR CANNOT PAY THE WAGES OF THE EMPLOYEES IT ENGAGED TO
PERFORM
THE JOB OR SERVICE, WILL THE PRINCIPAL AUTOMATICALLY BECOME THE
EMPLOYER
OF SUCH EMPLOYEES?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">No.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">Under
Article
106, a principal has two types of liability in relation to the
employees
of the subcontractor. The first type of liability is limited, and is
governed
by the first two paragraphs of Article 106. Thus, mere inability of the
subcontractor to pay wages will not automatically make the principal
the
direct employer. It will only make the principal jointly and severally
liable with the subcontractor for payment of the employees' wages to
the
extent of the work performed under the contract.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">The second
type
of liability, which arises from the third and fourth paragraphs of
Article
106, is absolute and direct. This liability arises when there is
labor-only
contracting as defined in <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a>. In such cases, the principal shall be responsible to the
workers in the same manner and extent as if it directly employed these
workers.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#003300">15. WHAT DOES
NON-IMPAIRMENT
OF EXISTING CONTRACTS MEAN AND WHY IS THIS NECESSARY?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Section
3 of <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> states that rights or benefits enjoyed by parties in
contracts
executed prior to <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> shall not be impaired. The contracts referred to are those
contracts executed and already being implemented before <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> took effect on 29 May 2001. Accordingly, the obligations,
rights and benefits or parties to any subcontracting arrangement prior
to the effectivity of <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> shall not be diminished, subject to Articles 106 to 109 of
the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>, and jurisprudence. The non-impairment provision in <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> is derived from the Constitutional principle against
non-impairment
of contracts.</font></font></strong>
</blockquote><strong>
<font face="Verdana"><font color="#003300">16. UNDER <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. NO. 10</a>, THERE WAS A REGISTRY OF SUBCONTRACTORS ESTABLISHED IN
DOLE.
WHAT IS THE EFFECT OF REVOCATION ON THIS REGISTRY?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">
<a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> abolished the DOLE registry of subcontractors. Thus, there
is no more requirement for subcontractors to register in DOLE.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">The
purpose
of the DOLE registry of subcontractors is specific. If a subcontractor
enrolls in this registry, it enjoys the presumption that it is engaged
in legitimate subcontracting. The burden of proving that it is an
illegitimate
or an illegal subcontractor will then be on the person claiming it.
With
the revocation, there is no more difference between DOLE-registered
subcontractors
and those that are not.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">Abolition
of
the DOLE registry, however, does not mean that a subcontractor will no
longer register at all. A subcontractor must still follow the
registration
or licensing procedures required in other applicable laws. For example,
a corporation or cooperative which seeks to operate as a subcontractor
should still register with the Securities and Exchange Commission or
the
Cooperative Development Authority, as the case may be. Likewise, the
abolition
of the DOLE registry does not exempt a subcontractor from the licensing
or permit requirements administered by relevant regulatory agencies.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#003300">17. <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. NO. 10</a> CONTAINED PROVISIONS ON SECURITY OF TENURE AND PROCEDURES
FOR DISMISSAL. HAVE THESE BEEN REVOKED BY <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. NO. 3</a>?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">
<a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O No. 10</a> was revoked in its entirety by <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a>. Thus, <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a> itself can no longer be cited as an implementing
guideline
of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a> provisions on security of tenure and dismissal of employees.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">However,
the
provisions of <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a> on security of tenure and dismissal are identical with
the
provisions of Rule XXIII, D. O. No. 9, series of 1997. These provisions
of D. O. No. 9 are not affected by <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a>, and, therefore, remain in force relative to security of
tenure
and employee dismissal.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#003300">18. AFTER THE
REVOCATION
OF <a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. NO. 10</a>, ARE THERE PLANS FOR THE FORMULATION OF NEW GUIDELINES TO
IMPLEMENT ARTICLES 106 TO 109?</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">Yes.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">
<a href="https://www.chanrobles.com/doledepartmentorderno10.html">D.
O. No. 10</a> was revoked to give government, workers and employers an
opportunity to formulate a new set of rules that is more responsive to
current employment arrangements and more acceptable to all concerned.
Accordingly,
Section 4 of <a href="https://www.chanrobles.com/doledepartmentorderno3.html">D.
O. No. 3</a> mandates that new guidelines shall be formulated by DOLE
upon
prior consultations with all sectors concerned, particularly the
Tripartite
Industrial Peace Council (TIPC) established under Executive Order No.
49,
series of 1998.</font></font></strong>
</p>
</blockquote>
</div>"""),
test.collect_content_individually_bs("""<center>
<a name="CHAN ROBLES AND ASSOCIATES LAW FIRM - PHILIPPINE LABOR CIRCULAR-ON-LINE">
</a>
<a name="TOP OF LABOR CODE PAGE">
</a>
<br>
<br>
<strong>
<font face="CG Times"><font color="#8b7b45"><font size="+1">
<img src="" style="width: 520px; height: 66px;"></font>
</font></font></strong>
<br>
<br>
<center><br><strong>
<font face="Verdana"><font color="#330000">This web page
features
the full text of</font></font></strong>
<br><strong>
<font face="Verdana"><font color="#330000">
<a href="#Department%20Order%20No.%2010">Department
Order No. 10 (Series of 1997)</a>.</font></font></strong>
<br><strong>
<font face="Verdana"><font color="#333300"><font size="-1">AMENDING
THE RULES IMPLEMENTING BOOKS III AND VI OF THE LABOR CODE, AS AMENDED.</font></font>
</font></strong></center>
<center>
<center>
<table style="" border="0" cellpadding="5" cellspacing="5" cols="1" width="45%">
<tbody>
<tr>
<td>
<center><strong>
<font face="Verdana"><font color="#524744"><font size="-1">Read
Full Text of the Rules on Contracting and Subcontracting Arrangement:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font>
</font></strong></center>
</td>
</tr>
<tr>
<td>
<center>
<li><strong>
<font color="#524744"><font size="-2">
<a href="https://www.chanrobles.com/doledepartmentorderno18-02-2002.html"><font face="Verdana">Department
Order No. 18-02 (Series of 2002) <em>[</em></font>
<em>
<font face="Verdana">Rules
Implementing Articles 106 to 109 of the Labor Code, as Amended]</font>
</em>
</a></font>
</font></strong>
</li></center>
</td>
</tr>
<tr>
<td>
<center>
<li><strong>
<font face="Verdana"><font color="#524744"><font size="-1">
<a href="https://www.chanrobles.com/doledepartmentorderno03-2001.html">Department
Order No. 03 (Series of 2001) <em>[Revoking Department Order No. 10,
Series
of 1997, and Continuing to Prohibit Labor-Only Contracting]</em>
</a></font>
</font></font></strong>
</li></center>
</td>
</tr>
<tr>
<td>
<center><strong>
<font face="Verdana"><font color="#330000"><font size="-2">
<a href="https://www.chanrobles.com/dolecontractingsubcontractingprimer.html" onmouseover="window.status='DOLE PRIMER ON CONTRACTING AND SUBCONTRACTING'; return true" onmouseout="window.status='';return true">DOLE
Primer on Contracting and Subcontracting Under Articles 106 to 109 of
the
Labor Code</a></font>
</font></font></strong></center>
</td>
</tr>
<tr>
<td>
<center><strong>
<font face="Verdana"><font size="-2">
<a href="https://www.chanrobles.com/doledepartmentorderno10-1997.htm" onmouseover="window.status='DEPARTMENT ORDER NO. 10, SERIES OF 1997'; return true" onmouseout="window.status='';return true"><font color="#330000">Department
Order No. 10 (Series of 1997) [</font>
<font color="#333300">Amending
the
Rules Implementing Books III and VI of the Labor Code, as Amended]</font>
</a></font>
</font></strong></center>
</td>
</tr>
</tbody>
</table></center>
</center>
<center>
<center>
<table border="0" cols="1" height="10%" width="30%">
<tbody>
<tr>
<td>
<center><strong>
<font face="Verdana"><font color="#ff6600"><font size="-2">
<a href="https://www.chanrobles.com/virtualibrary.htm" onmouseover="window.status='CHAN ROBLES VIRTUAL LAW LIBRARY '; return true" onmouseout="window.status='';return true">THE
CHAN ROBLES VIRTUAL LAW LIBRARY - <blink>QUICK</blink> GLANCE</a></font>
</font></font></strong></center>
</td>
</tr>
<tr>
<td>
<center>
<font size="-2"><strong>
<font face="Verdana">
<a href="https://www.chanrobles.com/virtualibrary1.htm" onmouseover="window.status='PHILIPPINE LEGAL RESOURCES - VIRTUAL LAW LIBRARY OF CHAN ROBLES'; return true" onmouseout="window.status='';return true"><font color="#8b7b45">Philippines</font>
</a>
<font color="#0000ff">&nbsp;</font></font></strong>
<font face="Verdana"><font color="#0000ff"></font>
</font>
<font color="#524744">|&nbsp;&nbsp;</font>
<font face="Verdana"><font color="#0000ff"></font><strong>
<font color="#8b7b45">
<a href="https://www.chanrobles.com/legal2ab.htm" onmouseover="window.status='WORLDWIDE LEGAL RESOURCES - VIRTUAL LAW LIBRARY OF CHAN ROBLES'; return true" onmouseout="window.status='';return true">Worldwide</a></font></strong>
<font color="#0000ff">&nbsp;&nbsp;
</font></font>
<font color="#524744">|</font>
<font face="Verdana"><font color="#0000ff">&nbsp;
</font><strong>
<font color="#8b7b45">
<a href="https://www.chanrobles.com/virtualibrary2.htm" onmouseover="window.status='THE BUSINESS PAGE - VIRTUAL LAW LIBRARY OF CHAN ROBLES'; return true" onmouseout="window.status='';return true">The
Business Page</a></font></strong></font>
</font></center>
</td>
</tr>
</tbody>
</table></center>
</center>
<center>
<br>
<a name="Department Order No. 10">
</a>
<br>
<br>
<strong>
<font face="Arial Black"><font color="#003300"><font size="-0">DEPARTMENT
ORDER NO. 10</font></font>
</font></strong>
<br><strong>
<font face="Verdana"><font color="#524744">(Series of 1997)</font></font></strong>
<br>
<br><strong>
<font face="Verdana"><font color="#333300"><font size="-0">AMENDING
THE RULES IMPLEMENTING BOOKS III AND VI</font></font>
</font></strong>
<br><strong>
<font face="Verdana"><font color="#333300"><font size="-0">OF
THE <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">LABOR
CODE, AS AMENDED</a></font>
</font></font></strong>
<br>
<font color="#333333">&nbsp; <br></font>
<div style="text-align: justify;"><font color="#333333">
\]&nbsp; </font>
<br>
</div></center>
<div style="text-align: justify;"><strong>
<font face="Verdana"><font color="#003300">ARTICLE I. There
is hereby issued a new Rule implementing Articles 106 to 109 of Book
III
of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>, to be known as Rule VIII-A, Book III of the Implementing
Rules,
as follows:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote>
<center><strong>
<font face="Verdana"><font color="#524744">"Rule
VIII-A</font></font></strong></center>
</blockquote>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"Section
1. <em>Guiding Principles</em>. - This Rule is being issued in
recognition
of the following guiding principles:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
Contracting and subcontracting arrangements are expressly allowed by
law,
but may be subject to regulations consistent wit the promotion of
employment,
protection of workers' welfare and enhancement of industrial peace and
rights of workers to self-organization and collective bargaining; for
this
reason, labor-only contracting as defined herein shall be prohibited.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b)
Contractors
and subcontractors as well as their employees, are entitled to all the
rights and privileges, and are subject to all the duties and
responsibilities
which the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code, as amended</a>, attaches to every employee-employer relationship;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c)
Flexibility
for the purpose of increasing efficiency and streamlining operations is
essential for every business to grow in an atmosphere of free
competition;
however, any form of flexibility intended to circumvent or evade
workers'
rights shall in no case be countenanced; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(d) The
establishment
of an effective labor market information system is indispensable in the
formulation of policies, strategies and programs for human resource
development
supportive of and responsive to the needs of workers and enterprises.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"Section 2. <em>Coverage</em>.
- This Rule shall apply to all parties of contracting and
subcontracting
arrangements where employee-employer relationship exists.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
3.
<em>Parties.
-</em> A contracting or subcontracting arrangement involves a trilateral
relationship under which there is a contract for a specific job,
service,
or work between the principal and the contractor or subcontractor, and
a contract of employment between the contractor or subcontractor and
its
workers. Therefore, the parties to a contracting or subcontracting
arrangement
shall be the principal, the contractor or subcontractor, and the
workers
engaged by the latter. The principal and the contractor or
subcontractor
may be a natural or juridical person.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
4.
<em>Definition
of Terms. -</em> As used in this Rule, the following shall mean:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
<em>"Principal"</em>
refers to any employer who puts out or farms out a job, service, or
work
to a contractor or subcontractor, whether or not the arrangement is
covered
by a written contract.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b) <em>"Contractor
or subcontractor"</em> refers to any person or entity engaged in a
legitimate
contracting or subcontracting arrangement as defined in paragraph (d)
hereof.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c) <em>"Contractual
employee"</em> includes one employed by a contractor or subcontractor to
perform or complete a job, work or service pursuant to an arrangement
between
the latter and a principal as defined in paragraph (d) hereof. The term
excludes employees of the contractor or subcontractor engaged to
perform
a job, work or service not within the scope of the contract between the
latter and a principal.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(d) <em>"Contracting"
</em>or
<em>"subcontracting"</em> refers to an arrangement whereby a
principal agrees
to put out or farm out with a contractor or subcontractor the
performance
or completion of a specific job, work or service within a definite or
predetermined
period, regardless of whether such job, work or service is to be
performed
or completed within or outside the premises of the principal as
hereinafter
qualified.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Subject
to
the provisions of Sections 6, 7 and 8 of this Rule, contracting or
subcontracting
shall be legitimate if the following circumstances concur:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(i)
The contractor or subcontractor carries on a distinct and independent
business
and undertakes to perform the job, work or service on its own account
and
under its own responsibility, according to its own manner and method,
and
free from the control and direction of the principal in all matters
connected
with the performance of the work except as to the results thereof;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(ii)
The contractor
or subcontractor has substantial capital or investment; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(iii)
The agreement
between the principal and contractor or subcontractor assures the
contractual
employees' entitlement to all labor and occupational safety and health
standards, free exercise of the right to self-organization, security of
tenure, and social and welfare benefits.</font></font></strong>
</p>
</blockquote>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"(e) <em>"Substantial
capital or investment"</em> refers to the adequacy of resources actually
and directly used by the contractor or subcontractor in the performance
or completion of the job, work or service contracted out. It may refer
to capital stocks and subscribed capitalization in the case of
corporations,
tools, equipment, implement, machineries, uniforms, protective gear, or
safety devices actually used in the performance of the job, work or
service
contracted out. It likewise includes operating costs, administrative
costs
such as training and overhead costs, and such expenses as are necessary
to enable the contractor or subcontractor to exercise control,
supervision
or direction over its employees in all aspects of performing or
completing
the job, service or work contracted out. The phrase, however, excludes
all capital and investment the contractor or subcontractor may have
which
are not actually and directly used in the conduct of its business, or
any
gratuitous assistance, financial or otherwise, it may have received
from
the principal.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(f) <em>"Labor-only
contracting"</em> prohibited under this Rule is an arrangement where the
contractor or subcontractor merely recruits, supplies or places workers
to perform a job, work or service for a principal and the following
elements
are present:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(i)
The contractor or subcontractor does not have substantial capital or
investment
to actually perform the job, work or service under its own account and
responsibility; and</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(ii)
The employees
recruited, supplied or placed by such contractor or subcontractor are
performing
activities which are directly related to the main business of the
principal.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"(g) <em>"In-house
agency"</em> refers to a contractor or subcontractor engaged in the
supply
of labor which;</font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(i)
Is owned, managed or controlled by the principal; and</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(ii)
Operates
solely for the principal owning, managing or controlling it.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"(h) <em>"Bureau"
</em>refers
to the Bureau of Local Employment of the Department of Labor and
Employment.<em>
"Regional Office"</em> refers to the offices of the Department
established
in each of the regions.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
5.
<em>Term
or duration of contractual employment. -</em> Subject to the provisions
of sections 6, 7 and 8 hereof, the term or duration of contractual
employment
shall be coextensive with the term or duration of the contract between
the principal and contractor or subcontractor. However, where the
contract
is divisible into phases such that substantially different skills are
required
for each phase, the term or duration of the contractual employment may
be made coextensive with each phase.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"For
purposes
of this Rule, he phrase <em>"substantially different skills"</em> refer
to
those skills the acquisition of which requires specialized knowledge or
training.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
6.
<em>Permissible
contracting or subcontracting</em>. - Subject to the conditions set
forth
in Section 3 (d) and (e) and Section 5 hereof, the principal may engage
the services of a contractor or subcontractor for the performance of
any
of the following;</font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
Works or services temporarily or occasionally needed to meet abnormal
increase
in the demand of products or services, provided that the normal
production
capacity or regular workforce of the principal cannot reasonably cope
with
such demands;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b)
Works or
services temporarily or occasionally needed by the principal for
undertakings
requiring expert or highly technical personnel to improve the
management
or operations of an enterprise;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c)
Services
temporarily needed for the introduction or promotion of new products,
only
for the duration of the introductory or promotional period;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(d)
Works or
services not directly related or not integral to the main business or
operation
of the principal, including casual work, janitorial, security,
landscaping,
and messengerial services, and work not related to manufacturing
processes
in manufacturing establishments;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(e)
Services
involving the public display of manufacturers' products which do not
involve
the act of selling or issuance of receipts or invoices;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(f)
Specialized
works involving the use of some particular, unusual or peculiar skills,
expertise, tools or equipment the performance of which is beyond the
competence
of the regular workforce or production capacity of the principal; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(g)
Unless
a reliever system is in place among the regular workforce, substitute
services
for absent regular employees, provided that the period of service shall
be coextensive with the period of absence and the same is made clear to
the substitute employee at the time of engagement. The phrase <em>"absent
regular employees"</em> includes those who are serving suspensions or
other
disciplinary measures not amounting to termination of employment meted
out by the principal, but excludes those on strike where all the formal
requisites for the legality of the strike have been <em>prima facie</em>
complied with based on the records filed with the National Conciliation
and Mediation Board.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"Section 7. <em>Prohibitions</em>.
- The following are hereby declared prohibited for being contrary to
law
or public policy;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(a)
Labor-only
contracting;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(b)
Contracting
out of work which will either displace employees of the principal from
their jobs or reduce their regular work hours;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c)
Contracting
out of work with a <em>"cabo"</em> as defined in Section 1 (ii), Rule I,
Book V of these Rules;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(d)
Taking
undue advantage of the economic situation of lack of bargaining
strength
of the contractual employee, or undermining his security of tenure or
basic
rights, or circumventing the provisions of regular employment, in any
of
the following instances:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(i)
In addition to this assigned functions, requiring the contractual
employee
to perform functions which are currently being performed by the regular
employees of the principal or of the contractor or subcontractor;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(ii)
Requiring
him to sign, as a precondition to employment or continued employment,
an
antedated resignation letter; a blank payroll; a waiver of labor
standards
including minimum wages and social or welfare benefits; or a quitclaim
releasing the principal, contractor or subcontractor from any liability
as to payment of future claims; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(iii)
Requiring
him to sign a contract fixing the period of employment to a term
shorter
than the term of the contract between the principal and the contractor
or subcontractor, unless the latter contract is divisible into phases
for
which substantially different skills are required and this is made
known
to the employee at the time of engagement;</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"(e)
Contracting
out of a job, work or service through an in-house agency as defined
herein;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(f)
Contracting
out of a job, work or service directly related to the business or
operation
of the principal by reason of a strike or lockout whether actual or
imminent;
and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(g)
Contracting
out of job, work or service when not justified by the exigencies of the
business and the same results in the reduction or splitting of the
bargaining
unit.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
8.
<em>Unfair
Labor Practice</em>. - Contracting out of a job, work or service being
performed
by union members when such will interfere with, restrain or coerce
employees
in the exercise of their rights to self-organization shall be unlawful
and shall constitute unfair labor practice.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
9.
<em>Contract
between contractor or subcontractor and contractual employee</em>. -
Notwithstanding
oral or written stipulations to the contrary, the contract between the
contractor or subcontractor and the contractual employee shall include
the following terms and conditions:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
The specific description of the job, work or service to be performed by
the contractual employee;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b) The
place
of work and terms and conditions of employment, including a statement
of
the wage rate applicable to the individual contractual employee; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c) The
term
or duration of employment, which shall be coextensive with the contract
between the principal and contractor or subcontractor, or with the
specific
phase for which the contractual employee is engaged, as the case may be.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">The
contractor
or subcontractor shall inform the contractual employee of the foregoing
terms and conditions on or before the first day of his employment.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
10.
<em>Duty
to produce copy of contract. -</em> The contractor or subcontractor
shall
submit a copy of its contract with the principal to the Regional Office
of the Department of Labor and Employment (DOLE). It shall be
accompanied
by a statement of the number of employees covered by the contract and,
where appropriate, a description of the phases of the contract and the
number of employees covered in each phase. The contractor or
subcontractor
shall be under an obligation to produce the original copy of the same
in
the ordinary course of inspection or when directed to do so by the
Regional
Director or his authorized representative.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"The copy
of
the contract between the contractual employee and the contractor or
subcontractor
need not be filed with DOLE. However, the contractor or subcontractor
shall
make the same available for inspection by the Regional Director or his
authorized representative.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Further,
a
copy of the contract between the contractual employee and the
contractor
or subcontractor shall be furnished the certified bargaining agent, if
there is any.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
11.
<em>Rights
of a contractual employee</em>. - The contractual employee shall be
entitled
to all the rights and privileges due a regular employee, including but
not limited to the right to working conditions and standards, service
incentive
leave, rest days, overtimes and holidays, health, safety and social and
welfare benefits, self-organization and collective bargaining and
security
of tenure.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"In cases
of
termination of employment prior to the expiration of the contract
between
the principal and the contractor or subcontractor, the right of the
contractual
employee to separation pay or other related benefits shall be governed
by the applicable laws and jurisprudence on termination of employment.
Where the termination results from expiration of the contract between
the
principal and the contractor or subcontractor, or whom the completion
of
the phase of the job, work or service for which the contractual
employee
is engaged, the latter shall not be entitled to separation pay.
However,
this shall be without prejudice to completion bonuses or other
emoluments,
including retirement pay as may be provided by law or in the contract
between
the principal and the contractor or subcontractor.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
12.
<em>Employee-employer
relationship. -</em> Except in cases provided for in Sections 13, 14, 15
and 17, the contractor or subcontractor shall be considered the
employer
of the contractual employee for purposes of enforcing the provisions of
the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Code</a>.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
13.
<em>When
principal is deemed jointly and severally liable</em>. - When the
contractor
or subcontractor fails to pay the wages of its contractual employees,
the
principal shall be jointly and severally liable with the contractor or
subcontractor to such contractual employees to the extent of the work
performed
under the contract, in the same manner and extent that the principal is
liable to its direct employees.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
14.
<em>When
principal is deemed employer who is solidarily liable. -</em> The
principal
shall be deemed as the direct employer of the contractual employees
and,
therefore, solidarily liable with the contractor or subcontractor for
whatever
monetary claims the contractual employees may have against the former,
in the following cases:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
When the contractor or subcontractor is not enrolled in the registry of
the Regional Office of the Bureau, or it has been delisted therefrom,
or
its contract with the principal has not been renewed;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b)
When the
contractor or subcontractor is found committing any of the prohibited
activities
enumerated in Section 7 of this Rule;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c)
When the
contractor or subcontractor is declared guilty of unfair labor practice
as specified in Section 8 of this Rule; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(d)
When a
violation of the relevant provisions of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Code</a>
has been established by the Regional Director in the exercise of his
enforcement
powers.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"Section 15. <em>Other
instances of solidary liability.</em> - In cases not covered by the last
two preceding sections, the principal shall also be deemed solidarily
liable
with the contractor or subcontractor to the extent of accrued claims
and
benefits which the latter may owe to its contractual employees in the
following
instances:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
When the certificate of registration, license or business permit of the
contractor or subcontractor is cancelled, revoked or not renewed by the
competent authority; or</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b)
When the
contract between the principal and the contractor or subcontractor is
preterminated
for reasons nor attributable to the fault of the contractor or
subcontractor.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"Section 16. <em>Enforcement
against performance bond</em>. - In enforcing the provisions of Sections
13, 14 and 15, the Regional Director shall first proceed against the
performance
bond, if any has been put up by the contractor or subcontractor, to the
extent that such bond may cover claims of the contractual employee. In
case the bond is insufficient, the Regional Director shall proceed
directly
against the principal.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"Nothing
herein
shall restrict the right of the principal from filing an action for
reimbursement
or damages against the contractor or subcontractor in the appropriate
courts.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
17.
<em>Effect
of existence of labor-only contracting.</em> - In case of declaration by
the competent authority that a contractual employee is covered by a
labor-only
contracting arrangement, he shall be considered part of the bargaining
unit of the principal.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
18.
<em>Effect
of expiration of contract; manpower pool</em>. - Where the contract
between
the principal and the contractor or subcontractor has expired, and the
latter remains in business as a contractor or subcontractor, the
employee-employer
relationship between the latter and its contractual employees shall not
be automatically terminated, but shall remain suspended for a period of
six months, unless a longer period is set by the contractor or
subcontractor.
During this period, such employees shall become part of a manpower pool
of the contractor or subcontractor. If the contractor or subcontractor
is unable to renew the original contract or enter into a new and
similar
contract requiring the skills of the employees in the pool within the
six-month
period, or if the contractual employee subsequently finds employment
elsewhere,
the employee-employer relationship shall be deemed terminated and the
employee
concerned shall be taken out of the pool.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"The
foregoing
paragraph shall also apply where the contract between the contractor or
subcontractor and the contractual employee has expired by reason of the
completion of the phase of the contract for which the latter was
engaged.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"It shall
be
understood, however, that all rights and privileges which the employee
may derived out of the employer-employee relationship shall be
suspended
while he is part of the pool.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
19.
<em>Registry
of contractors or subcontractors.</em> - There is hereby established a
registry
of contractors and subcontractors in the Regional Offices and in the
Bureau,
for purposes of establishing an effective labor market information and
monitoring system on activities which are subject to contracting or
subcontracting
arrangements. Registration under this section shall not be synonymous
with
licensing, the latter being a precondition for acquiring legal
personality
or engaging in business.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
20.
<em>Requirements
for registration</em>. - A contractor or subcontractor shall be enrolled
in the registry of contractors and subcontractors upon completion of an
application form to be provided by the DOLE. The application shall
state:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
The name and business address of the applicant and the area or areas
where
it seeks to operation;</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b) The
names
and addresses of its officers, if the applicant is a corporation or
partnership;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c) The
nature
of the applicant's business and the industry or industries where the
applicant
seeks to operate;</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(d) The
list
of actual contracts, if any; and</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(e) The
capitalization
and other assets of the applicants which are actually and directly used
in its operations.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"The
application
shall be supported by:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
A certified copy of the certificate of registration of firm or business
name from the Securities and Exchange Commission (SEC) or Department of
Trade and Industry (DTI) or from the DOLE if the applicant is a union;
and</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b) A
certified
copy of the license or business permit issued by the local government
unit
or units where the contractor or subcontractor operates.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"The
application
shall be verified and shall include an undertaking that the contractor
or subcontractor shall abide by all applicable labor laws and
regulations.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
21.
<em>Filing
and processing of applications</em>. - The application and its
supporting
documents shall be filed in triplicate in the Regional Office where the
applicant principally operates. No application for registration shall
be
accepted unless all the foregoing requirements are complied with. The
contractor
or subcontractor shall be deemed registered upon payment of a
registration
fee of One hundred pesos (P100.00) to the Regional Office.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Where all
the
supporting documents have been submitted, the Regional Office shall
deny
or approve the application within seven (7) working days after its
filing.
In case of inaction of the Regional Office beyond this period, the
application
shall be deemed provisionally approved subject, however, to the payment
of the registration fee.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Upon
registration,
the Regional Office shall return one set of the duly-stamped
application
documents to the applicant, retain one set for its file, and transmit
the
remaining set to the Bureau. The Bureau shall devise the necessary
forms
for the expeditious processing of all applications for registration.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
22.
<em>Annual
reporting. -</em> The contractor or subcontractor shall submit in
triplicate
its annual report in such forms as may be prescribed by the DOLE to the
appropriate Regional Office. The report shall include:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
A list of contracts entered into during the subject reporting period;
and</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b) A
certification
from the Social Security System (SSS) and the Home Development Mutual
Fund
(HDMF) that the contractor or subcontractor has been making the monthly
remittances due its contractual employees during the subject reporting
period.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"The
obligation
to submit an annual report shall coincide with the anniversary date of
registration of the contractor or subcontractor. The Regional Office
shall
return one set of the duly-stamped report to the contractor or
subcontractor,
retain one set for its file, and transmit the remaining set to the
Bureau
within five (5) days from receipt thereof.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
23.
<em>Delisting
of registered contractor or subcontractor</em>. - The Regional Director
shall, upon due notice, <em>motu proprio</em> cancel the registration of
a contractor or subcontractor if it fails to comply with the reporting
requirements for three consecutive years, or upon the cessation of a
business
of the latter.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Subject
to
administrative due process, the contractor or subcontractor shall be
delisted
from the registry if it is found to have committed the prohibited
activities
or has been declared guilty of unfair labor practice as enumerated in
Sections
7 and 8 hereof, or has falsified the requirements for registration it
submitted
to the Regional Office.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
24.
All existing contractors or subcontractors as defined herein shall
register
with the Regional Office within one hundred twenty (120) days from the
effectivity of this Rule. In case of failure to register within this
prescribed
period, the provisions of this Rule shall apply.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Section
25.
<em>Supersession</em>.
- All rules and regulations issued by the Secretary of Labor and
Employment
inconsistent with the provisions of this Rule are hereby superseded.
Contracting
or subcontracting arrangements in the construction industry, however,
shall
continue to be governed by Department Order No. 19, series of 1993, as
well as the applicable provisions of this Rule."</font></font></strong>
</p>
</blockquote><strong>
<font face="Arial Black"><font color="#003300">ARTICLE II. Sections
7, 8 and 9, Rule VIII, Book III of the Implementing Rules are hereby
superseded.
Sections 10, 11, 12, 13 and 14 of the same Rule are hereby renumbered
as
Sections 7, 8, 9, 10 and 11, respectively.</font></font></strong>
</div>
<p style="text-align: justify;"><strong>
<font face="Arial Black"><font color="#003300">ARTICLE III. Section
2, Rule I, Book VI of the Implementing Rules is hereby amended, to read
as follows:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<div style="text-align: justify;">
<blockquote><strong>
<font face="Verdana"><font color="#524744">"Section
2. <em>Security of tenure. -</em> (a) In cases of regular employment,
the
employer shall not terminate the services of an employee except for
just
or authorized causes as provided by law, and subject to the
requirements
of due process.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b) The
foregoing
shall also apply in cases of probationary employment: <em>Provided,
however</em>,
that in such cases, termination of employment due to failure of the
employee
to qualify in accordance with the standards of the employer made known
to the former at the time of engagement may also be a ground for
termination
of employment.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c) In
cases
of employment covered by contracting or subcontracting arrangements, no
employee shall be dismissed prior to the expiration of the contract
between
the principal and contractor or subcontractor as defined in Rule
VIII-A,
Book III of these Rules, unless the dismissal is for just or authorized
cause, or is brought about by the completion of the phase of the
contract
for which the employee was engaged but, in any case, subject to the
requirements
of due process or prior notice.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(d) In
all
cases of termination of employment, the following standards of due
process
shall be substantially observed:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"For
termination of employment based on just cases as defined in Article 282
of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(i)
A written notice served on the employee specifying the ground or
grounds
for termination, and giving said employee reasonable opportunity within
which to explain his side.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(ii)
A hearing
or conference during which the employee concerned, with the assistance
of counsel, if he so desires, is given opportunity to respond to the
charge,
present his evidence, or rebut the evidence presented against him.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(iii)
A written
notice of termination served on the employee, indicating that upon due
consideration of all the circumstances, grounds have been established
to
justify his termination.</font></font></strong>
</p>
</blockquote><strong>
<font face="Verdana"><font color="#524744">"For
termination
of employment as defined in Article 283 of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a>, the requirement of due process shall be deemed complied with
upon service of a written notice to the employee and the appropriate
Regional
Office of the Department of Labor and Employment at least thirty days
before
effectivity of the termination, specifying the ground or grounds for
termination.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"If the
termination
is brought about by the completion of a contract or phase thereof, or
by
failure of an employee to meet the standards of the employer in the
case
of probationary employment, it shall be sufficient that a written
notice
is served the employee within a reasonable time from the effective date
of termination."</font></font></strong>
</p>
</blockquote>
</blockquote><strong>
<font face="Arial Black"><font color="#003300">ARTICLE IV. Section
5,
Rule I, Book VI of the Implementing Rules of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a> is hereby amended, to read as follows:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"Section
5. (a) <em>Regular employment. -</em> The provisions of written
agreements
to the contrary notwithstanding and regardless of the oral agreements
of
the parties, employment shall be deemed regular for purposes of Book VI
of the <a href="https://www.chanrobles.com/legal4labor.htm#LABOR%20CODE%20OF%20THE%20PHILIPPINES%20%5BFULL%20TEXT%5D">Labor
Code</a> where employee has been engaged to perform activities which
are
usually necessary or desirable in the usual business or trade of the
employer,
except where the employment has been fixed for a specific project or
undertaking,
the completion or termination of which has been determined at the time
of the engagement of the employee or where the job, work or service to
be performed is seasonal in nature and the employment is for the
duration
of the season.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b) <em>Casual
employment.</em> - There is casual employment where an employee is
engaged
to perform a job, work or service which is merely incidental to the
business
of the employer, and such job, work or service is for a definite period
made known to the employee at the time of engagement: <em>Provided</em>,
that any employee who has rendered at least one year of service,
whether
such service is continuous or not, shall be considered a regular
employee
with respect to the activity in which he is employed and his employment
shall continue while such activity exists.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"Notwithstanding
the foregoing distinctions, every employee shall be entitled to the
rights
and privileges, and shall be subject to the duties and obligations, as
may be granted by law to regular employees during the period of their
actual
employment."</font></font></strong>
</p>
</blockquote><strong>
<font face="Arial Black"><font color="#003300">ARTICLE V. Section 6,
Rule I, Book VI of the Implementing Rules is hereby amended, to read as
follows:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"Section
6. <em>Probationary employment</em>. - There is probationary employment
where
the employee, upon his engagement, is made to undergo a trial period
during
which the employer determines his fitness to qualify for regular
employment
based on reasonable standards made known to him at the time of
engagement.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"Probationary
employment shall be governed by the following rules:<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtuallawlibrary</span></font></font></strong>
</p>
<blockquote><strong>
<font face="Verdana"><font color="#524744">"(a)
Where the work for which the employee has been engaged is learnable or
apprenticeable in accordance with the standards prescribed by the
Department
of Labor and Employment, the period of probationary employment shall be
limited to the authorized learnership or apprenticeship period, which
is
applicable.</font></font></strong>
<p><strong>
<font face="Verdana"><font color="#524744">"(b)
Where the
work is neither learnable nor apprenticeable, the period of
probationary
employment shall not exceed six months reckoned from the date the
employee
actually started working.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(c) The
services
of an employee who has been engaged on probationary basis may be
terminated
only for a just or authorized cause, when he fails to qualify as a
regular
employee in accordance with reasonable standards prescribed by the
employer.</font></font></strong>
</p>
<p><strong>
<font face="Verdana"><font color="#524744">"(d) In
all
cases of probationary employment, the employer shall make known to the
employee the standards under which he will qualify as a regular
employee
at the time of his engagement. Where no standards are made known to the
employee at that time, he shall be deemed a regular employee."</font></font></strong>
</p>
</blockquote>
</blockquote><strong>
<font face="Arial Black"><font color="#003300">ARTICLE VI. <em>Effectivity</em>.
- These Rules shall take effect fifteen (15) days after completion of
its
publication in two (2) newspapers of general circulation.</font></font></strong>
</div>
<div align="right"><strong>
<font face="Verdana"><font color="#524744">Manila,
Philippines, 30 May 1997.</font></font></strong>
</div>
<center>
<br>
<br><strong>
<font face="Verdana"><font color="#003300">LEONARDO A.
QUISUMBING</font></font></strong>
<br><strong>
<font face="Verdana"><font color="#003300">Secretary</font></font></strong>
<br>
<br>
<font face="Verdana"><font color="#333333"><font size="-2">
<a href="#CHAN%20ROBLES%20AND%20ASSOCIATES%20LAW%20FIRM%20-%20PHILIPPINE%20LABOR%20CIRCULAR-ON-LINE">Back
to Top</a></font>
</font></font>&nbsp;&nbsp; -&nbsp; <font face="Verdana"><font color="#333333"><font size="-2">
<a href="https://www.chanrobles.com">Back
to Home</a></font>
</font></font>&nbsp;&nbsp; -&nbsp; <font face="Verdana"><font color="#333333"><font size="-2">
<a href="https://www.chanrobles.com/legal4.htm">Back
to Philippine Labor Circular Index</a></font>
</font></font>
<br>
<br>
<center><strong>
<a name="CHAN ROBLES &amp; ASSOCIATES LAW FIRM-BOTTOM">
</a></strong></center>
<center><strong>
<tt>
<font color="#333333"><font size="-2">&nbsp;<span style="color: #dadad2; font-size: 1pt;">chanroblesvirtualawlibrary</span><br>
<br>
&nbsp;&nbsp; <br></font></font></tt></strong></center>
</center>
<p><br>
</p> <br>
<p></p>
</center>""")
]
```
